### 1.

This is information of students participating in various games, represent it by a horizontal Bar graph.

| Game | Number of students |
| --- | --- |
| Cricket | 125 |
| Football | 59 |
| Hockey | 28 |
| Chess | 19 |
| Tennis | 36 |



```python
import matplotlib.pyplot as plt

games = ["Cricket", "Football", "Hockey", "Chess", "Tennis"]
participants = [125, 59, 28, 19, 36]

plt.figure(figsize=(8, 5))
bars = plt.barh(games, participants, color=['#1f77b4', '#2ca02c', '#ff7f0e', '#9467bd', '#8c564b'])
plt.xlabel('Number of Students')
plt.title('Students Participating in Different Games')
plt.xlim(0, max(participants) + 20)

for bar, value in zip(bars, participants):
    plt.text(value + 2, bar.get_y() + bar.get_height() / 2, str(value), va='center')

plt.tight_layout()
plt.show()
```

### 2.

Plot the graph of f(x) = sin(x) +cos(x) , for x  [–3 π, 3 π] in dashed dotted pattern.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3 * np.pi, 3 * np.pi, 1000)
y = np.sin(x) + np.cos(x)

plt.figure(figsize=(9, 4))
plt.plot(x, y, '-.', color='teal', linewidth=2, label=r'$f(x) = \sin(x) + \cos(x)$')
plt.title(r'Graph of $f(x) = \sin(x) + \cos(x)$ on $[-3\pi, 3\pi]$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True, linestyle=':')
plt.show()
```

### 3.

Plot 3D surface plot for the function f(x, y) = x4 – y2 for –5  x, y  5, in black .


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)
Z = X**4 - Y**2

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='black', edgecolor='none', alpha=0.9)
ax.set_title(r'Surface: $f(x, y) = x^4 - y^2$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```

### 4.

Using sympy declare the points P(7, 3), Q(7,−5) and R(7,1), check whether these points are
collinear. Declare the segment passing the points P and Q, find the midpoint of segment PQ.


```python
from sympy import Point, Segment

P = Point(7, 3)
Q = Point(7, -5)
R = Point(7, 1)

are_collinear = Point.are_collinear(P, Q, R)
print(f'Are points P{tuple(P)} , Q{tuple(Q)} , R{tuple(R)} collinear? {are_collinear}')

segment_pq = Segment(P, Q)
print(f'Midpoint of segment PQ: {segment_pq.midpoint}')
```

### 5.

Using sympy declare  ABC where A (2 , 2), B(4, 2), C( 3, 6) ,for this triangle find its centroid
and also translate ABC in x direction by 2 units.


```python
from sympy import Point, Triangle

A = Point(2, 2)
B = Point(4, 2)
C = Point(3, 6)

triangle = Triangle(A, B, C)
centroid = triangle.centroid
print(f'Centroid of triangle ABC: {centroid}')

translated_points = [pt.translate(2, 0) for pt in (A, B, C)]
print('Translated triangle (by 2 units along +x):')
for label, pt in zip(['A'', 'B'', 'C''], translated_points):
    print(f'  {label}: {pt}')
```

### Q2.1 (Attempt Any One)

1. Write a python program to perform each of the following transformations on the point
(5, −8). Plot the point and its following transformations using different colors and markers.
a) Refection through Y−axis,
b) Scaling in x−coordinate by factor 4.
𝜋
c) Rotation about origin through an angle .
2

```python
import numpy as np
import matplotlib.pyplot as plt

original = np.array([5.0, -8.0])

reflected = np.array([-original[0], original[1]])
scaled = np.array([4 * original[0], original[1]])
theta = np.pi / 2
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])
rotated = rotation_matrix @ original

plt.figure(figsize=(6, 6))
plt.scatter(*original, color='black', marker='o', label='Original (5, -8)')
plt.scatter(*reflected, color='red', marker='s', label='Reflection in Y-axis')
plt.scatter(*scaled, color='blue', marker='^', label='Scaling x by 4')
plt.scatter(*rotated, color='green', marker='D', label='Rotation by $\pi/2$')

plt.axhline(0, color='gray', linewidth=0.7)
plt.axvline(0, color='gray', linewidth=0.7)
plt.legend(loc='upper left')
plt.title('Transformations of Point (5, -8)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle=':')
plt.show()
```

2. Write python program to generate and plot 50 points on parabola y2 = 20 x , using iterative
formulae for 0  y  30.

```python
import matplotlib.pyplot as plt

num_points = 50
max_y = 30
step = max_y / (num_points - 1)

xs = []
ys = []
y_val = 0.0
for _ in range(num_points):
    x_val = (y_val ** 2) / 20
    xs.append(x_val)
    ys.append(y_val)
    y_val += step

plt.figure(figsize=(7, 4))
plt.plot(xs, ys, 'o', color='navy', label=r'$y^2 = 20x$')
plt.title('Points on the Parabola $y^2 = 20x$ for $0 \leq y \leq 30$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```
