Write python program to generate and plot graph of function 𝑓(𝑥) = 𝑥 3 for 𝑥 ∈ [−5,5 ]
and 𝑔(𝑥) = 𝑒 𝑥 for 𝑥 ∈ [1,10] in one graph window using 2 ×1 subplots.


```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-5, 5, 400)
f_x = x1**3

x2 = np.linspace(1, 10, 400)
g_x = np.exp(x2)

fig, axes = plt.subplots(2, 1, figsize=(7, 8), sharex=False)
axes[0].plot(x1, f_x, color="navy", linewidth=2)
axes[0].set_title(r"$f(x) = x^3$ on $[-5, 5]$")
axes[0].set_xlabel("x")
axes[0].set_ylabel("f(x)")
axes[0].grid(True, linestyle=":")

axes[1].plot(x2, g_x, color="darkgreen", linewidth=2)
axes[1].set_title(r"$g(x) = e^x$ on $[1, 10]$")
axes[1].set_xlabel("x")
axes[1].set_ylabel("g(x)")
axes[1].grid(True, linestyle=":")

plt.tight_layout()
plt.show()
```

### 3.

Draw Wireframes plot for the function f(x, y) = cos (x2+y2) when 0 x, y 2π,
with brown colour.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

grid = np.linspace(0, 2 * np.pi, 150)
X, Y = np.meshgrid(grid, grid)
Z = np.cos(X**2 + Y**2)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.plot_wireframe(X, Y, Z, color="saddlebrown", linewidth=0.7)
ax.set_title(r"Wireframe of $f(x, y) = \cos(x^2 + y^2)$")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
plt.show()
```

Using sympy declare 3D Point P(2 , –1 , 3) and rotate about z-axis through angle 900 ,


```python
from sympy import Matrix, pi, cos, sin

point = Matrix([2, -1, 3])
theta = pi / 2
rotation_z = Matrix(
    [
        [cos(theta), -sin(theta), 0],
        [sin(theta), cos(theta), 0],
        [0, 0, 1],
    ]
)

rotated_point = rotation_z * point

print(f"Original point: ({point[0]}, {point[1]}, {point[2]})")
print(f"Point after 90° rotation about z-axis: ({rotated_point[0]}, {rotated_point[1]}, {rotated_point[2]})")
```

1. Write a python program to plot the  ABC with vertices A(–2 , 7), B(–5 , –1)
and C(4, 3)
On this polygon apply 1) shearing in x-direction proportional to y by factor 3 and y-direction
proportional to x by factor 4, 2) reflection through x-axis, 3) translation in y direction by
factor 7 units. Plot original and transformed figures in different colour .

```python
import numpy as np
import matplotlib.pyplot as plt

vertices = np.array([[-2, 7], [-5, -1], [4, 3]])

def close_shape(points):
    return np.vstack([points, points[0]])

original = close_shape(vertices)

shear_x = np.array([[1, 3], [0, 1]])
shear_y = np.array([[1, 0], [4, 1]])
reflected = np.array([[1, 0], [0, -1]])
translation = np.array([0, 7])

shear_xy = (shear_x @ vertices.T).T
shear_xy = (shear_y @ shear_xy.T).T
shear_xy_closed = close_shape(shear_xy)

reflected_vertices = (reflected @ shear_xy.T).T
reflected_closed = close_shape(reflected_vertices)

translated_vertices = reflected_vertices + translation
translated_closed = close_shape(translated_vertices)

fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(original[:, 0], original[:, 1], label="Original", linewidth=2)
ax.plot(shear_xy_closed[:, 0], shear_xy_closed[:, 1], label="After X & Y Shear", linewidth=2)
ax.plot(reflected_closed[:, 0], reflected_closed[:, 1], label="After Reflection (x-axis)", linewidth=2)
ax.plot(translated_closed[:, 0], translated_closed[:, 1], label="After Translation", linewidth=2)

ax.set_aspect("equal")
ax.set_title("Transformations on Triangle ABC")
ax.legend()
ax.grid(True, linestyle=":")
plt.show()
```

2. Write a python program to generate and plot 75 points on circle x2 + y2 = 64 using iterative
formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

radius = 8
num_points = 75
theta = 2 * np.pi / num_points
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

points = []
vector = np.array([radius, 0.0])
for _ in range(num_points):
    points.append(vector.copy())
    vector = rotation_matrix @ vector

points = np.array(points)

plt.figure(figsize=(6, 6))
plt.scatter(points[:, 0], points[:, 1], color="indianred", s=25, label="Circle points")
plt.gca().add_patch(plt.Circle((0, 0), radius, fill=False, linestyle="--", color="gray"))
plt.title(r"Points on $x^2 + y^2 = 64$")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.grid(True, linestyle=":")
plt.legend()
plt.show()
```
