### 1.

Represent the following information using a bar graph (in brown colour ) ,

| Subject | Marks |
| --- | --- |
| Mathematics | 68 |
| English | 45 |
| Biology | 79 |
| Physics | 56 |
| Chemistry | 70 |



```python
import matplotlib.pyplot as plt

subjects = ["Mathematics", "English", "Biology", "Physics", "Chemistry"]
marks = [68, 45, 79, 56, 70]

plt.figure(figsize=(8, 5))
plt.bar(subjects, marks, color="saddlebrown")
plt.title("Marks Obtained per Subject")
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.ylim(0, 100)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

### 2.

Plot the graph of y = x3 +10 x – 5 , for x  [–10, 10] in red colour with dashed lines.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 500)
y = x**3 + 10 * x - 5

plt.figure(figsize=(8, 5))
plt.plot(x, y, color="red", linestyle="--", linewidth=2)
plt.title(r"Graph of $y = x^3 + 10x - 5$")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.grid(True, linestyle=":", alpha=0.5)
plt.show()
```

### 3.

Draw surface plot for the function f(x, y) = x2 + y2 when –10 x, y 10,
with pink colour.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

grid = np.linspace(-10, 10, 200)
X, Y = np.meshgrid(grid, grid)
Z = X**2 + Y**2

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, color="pink", edgecolor="none", alpha=0.9)
ax.set_title(r"Surface Plot of $f(x, y) = x^2 + y^2$")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
plt.show()
```

### 4.

Using sympy declare points P(7,2) , Q(1,8). Find the distance between them. Declare the line
passing through them. Find the slope and equation of the line passing through points P and Q.


```python
from sympy import Point, Line

P = Point(7, 2)
Q = Point(1, 8)
line_pq = Line(P, Q)

distance = P.distance(Q)
slope = line_pq.slope
equation = line_pq.equation()

print(f"P: {P}, Q: {Q}")
print(f"Distance between P and Q: {distance.evalf()}")
print(f"Slope of PQ: {slope}")
print(f"Equation of PQ: {equation}")
```

### 5.

Using sympy declare  ABC where A (2 , 2), B(4, 2), C( 3, 6) ,for this triangle find its centroid
and also translate  ABC in x direction by 2 units.


```python
from sympy import Point, Polygon

A = Point(2, 2)
B = Point(4, 2)
C = Point(3, 6)
triangle = Polygon(A, B, C)

centroid = triangle.centroid
translated_triangle = triangle.translate(2, 0)

print(f"Triangle ABC vertices: {triangle.vertices}")
print(f"Centroid of triangle ABC: {centroid}")
print(f"Translated triangle vertices (shifted +2 along x): {translated_triangle.vertices}")
```

### Q2.1 (Attempt Any One)

1. Write a python program to plot the polygon with vertices at P(4 , 3),Q(6, 3),R(6, 5), and
S(4, 5). Plot its reflections through, 1) x-axis, 2) y-axis . Also plot x axis and y axis in
black colours . All figures must be in different colours .Use labels and legend .

```python
import numpy as np
import matplotlib.pyplot as plt

vertices = np.array([[4, 3], [6, 3], [6, 5], [4, 5]])

def close_shape(points):
    return np.vstack([points, points[0]])

original = close_shape(vertices)
reflect_x = original * np.array([1, -1])
reflect_y = original * np.array([-1, 1])

fig, ax = plt.subplots(figsize=(7, 7))
ax.plot(original[:, 0], original[:, 1], color="blue", linewidth=2, label="Original")
ax.plot(reflect_x[:, 0], reflect_x[:, 1], color="green", linewidth=2, label="Reflection: x-axis")
ax.plot(reflect_y[:, 0], reflect_y[:, 1], color="magenta", linewidth=2, label="Reflection: y-axis")

ax.axhline(0, color="black", linewidth=1, label="x-axis")
ax.axvline(0, color="black", linewidth=1, label="y-axis")
ax.set_aspect("equal")
ax.set_title("Polygon and its Reflections")
ax.legend(loc="best")
ax.grid(True, linestyle=":", alpha=0.4)
plt.show()
```

2. Write a python program to generate and plot 100 points on circle x2+y2 = 25 using iterative
formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

radius = 5
num_points = 100
theta = 2 * np.pi / num_points
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

points = []
vector = np.array([radius, 0.0])
for _ in range(num_points):
    points.append(vector.copy())
    vector = rotation_matrix @ vector

points = np.array(points)

plt.figure(figsize=(6, 6))
plt.plot(points[:, 0], points[:, 1], "o", color="purple", markersize=4, label="Circle points")
circle = plt.Circle((0, 0), radius, color="gray", fill=False, linestyle="--", alpha=0.5)
plt.gca().add_patch(circle)
plt.title(r"Points Generated on $x^2 + y^2 = 25$")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.grid(True, linestyle=":")
plt.legend()
plt.show()
```
