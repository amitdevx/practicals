### 4.

Using sympy declare points P ( 0, 0 ) , Q (1 , 1 ), R (– 7 ,5) . Find the distance of points Q and
R from P. Also find midpoint of line segment QR.


```python
import sympy as sp
P = sp.Point(0, 0)

Q = sp.Point(1, 1)

R = sp.Point(-7, 5)

d_PQ = P.distance(Q)

d_PR = P.distance(R)

mid_QR = Q.midpoint(R)

print(f"Distance PQ: {d_PQ}")

print(f"Distance PR: {d_PR}")

print(f"Midpoint QR: {mid_QR}")
```

### 5.

Using sympy declare polygon ABCD where A (1 , 1), B (–1, 1 ), C ( –1, –1 ) and D ( 1 , –1) ,
𝜋
Find angles of polygon ABCD at each vertices. Also find its rotation about origin by radians.
3


```python
import sympy as sp
A, B, C, D = sp.Point(1, 1), sp.Point(-1, 1), sp.Point(-1, -1), sp.Point(1, -1)

poly = sp.Polygon(A, B, C, D)

print(f"Angles: {poly.angles}")

theta = sp.pi / 3

rotated_verts = []

for p in poly.vertices:
    rotated_verts.append(p.rotate(theta, sp.Point(0,0)))

print("Rotated Vertices (pi/3):", rotated_verts)
```

### Q2.1 (Attempt Any One)

1. Write a python program to plot the Triangle with vertices at P (1 , 3), Q (3 , 3), R(4 , 5) and its
reflections through, 1) line y = x 2) line y = – x . All the figures must be in different colours.

```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
def apply_transformation(points, matrix):
    return np.dot(points, matrix.T)

poly = np.array([[1, 3], [3, 3], [4, 5]])

poly_plot = np.vstack([poly, poly[0]])

Ref1 = np.array([[0, 1], [1, 0]])

Ref2 = np.array([[0, -1], [-1, 0]])

p1 = apply_transformation(poly, Ref1)

p1_plot = np.vstack([p1, p1[0]])

p2 = apply_transformation(poly, Ref2)

p2_plot = np.vstack([p2, p2[0]])

plt.figure()

plt.plot(poly_plot[:, 0], poly_plot[:, 1], 'b-', label='Original')

plt.plot(p1_plot[:, 0], p1_plot[:, 1], 'g--', label='Refl y=x')

plt.plot(p2_plot[:, 0], p2_plot[:, 1], 'r--', label='Refl y=-x')

plt.legend()

plt.grid(True)

plt.axis('equal')

plt.title("Triangle Reflections")

plt.show()
```

2. Write a python program to generate and plot 50 points on circle x2 +y2 = 16 using iterative
formulae.

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

# Parameters
r_sq = 16
r = np.sqrt(r_sq)
num_points = 50

# Generate points using iterative formula (parametric approach)
theta = np.linspace(0, 2*np.pi, num_points)
x = r * np.cos(theta)
y = r * np.sin(theta)

# Plot
plt.figure(figsize=(6,6))
plt.plot(x, y, label=f'x^2 + y^2 = {r_sq}')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Circle: x^2 + y^2 = {r_sq}')
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```
