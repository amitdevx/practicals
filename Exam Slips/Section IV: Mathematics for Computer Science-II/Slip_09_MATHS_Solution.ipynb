Write python program to plot the graph of function 𝑓(𝑥) = 𝑠𝑖𝑛 (x 2 ) for 𝑥 ∈ [−10,10] and
𝑔(𝑥) = cos (x 2 ) for 𝑥 ∈ [0, 𝜋] in one graph window 2 × 1 subplots.


```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.linspace(-10, 10, 100)

y1 = np.sin(x1**2)

x2 = np.linspace(0, np.pi, 100)

y2 = np.cos(x2**2)

plt.figure()

plt.subplot(2, 1, 1)

plt.plot(x1, y1)

plt.title('f(x) = sin(x^2)')

plt.subplot(2, 1, 2)

plt.plot(x2, y2)

plt.title('g(x) = cos(x^2)')

plt.tight_layout()

plt.show()
```

### 2.

Plot the graph of 𝑦 = 𝑥 2 + 𝑒 2𝑥+5 , for x  [–10,10] in red color with dashed dotted line.


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5, 2, 50)

y = x**2 + np.exp(2*x + 5)

plt.figure()

plt.plot(x, y, 'r-.', label='y = x^2 + e^(2x+5)')

plt.xlabel('x')

plt.ylabel('y')

plt.legend()

plt.grid(True)

plt.show()
```

### 3.

Draw Surface plot for the function f(x, y) = 𝑥 3 + 𝑦 3 , when –20 x, y 20,
with pink colour.


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-20, 20, 30)

y = np.linspace(-20, 20, 30)

X, Y = np.meshgrid(x, y)

Z = X**3 + Y**3

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, color='pink')

ax.set_title('Surface Plot x^3 + y^3')

plt.show()
```

1. Write a python program to plot the polygon ABCD with vertices A(4, 4), B(–4, 4), C(–4, –4)
and D(4 , –4). Apply the following sequence of combined transformations on it
1) shearing in y-direction by factor –3, 2) reflection through x-axis, 3) translation x and y
direction by factors 5 and 8 respectively. And plot transformed polygon in different colours.

```python
import matplotlib.pyplot as plt
import numpy as np
def apply_transformation(points, matrix):
    return np.dot(points, matrix.T)

poly = np.array([[4, 4], [-4, 4], [-4, -4], [4, -4]])

poly_plot = np.vstack([poly, poly[0]])

Sh = np.array([[1, 0], [-3, 1]])

Ref = np.array([[1, 0], [0, -1]])

Trans = np.array([5, 8])

p1 = apply_transformation(poly, Sh)

p2 = apply_transformation(p1, Ref)

p3 = p2 + Trans

p3_plot = np.vstack([p3, p3[0]])

plt.figure()

plt.plot(poly_plot[:, 0], poly_plot[:, 1], 'b-', label='Original')

plt.plot(p3_plot[:, 0], p3_plot[:, 1], 'r-', label='Transformed (Final)')

plt.legend()

plt.grid(True)

plt.axis('equal')

plt.title("Combined Transformations")

plt.show()
```

2. Write a python program to generate and plot 75 points on parabola y2 = 36x
where 0 ≤ 𝑥 ≤ 30 using iterative formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
k = 36 # from y^2 = 36x
num_points = 75
x_min, x_max = 0, 10

# Generate points
x = np.linspace(x_min, x_max, num_points)
y_pos = np.sqrt(k * x)
y_neg = -np.sqrt(k * x)

# Plot
plt.figure()
plt.plot(x, y_pos, 'b-', label=f'y^2 = {k}x')
plt.plot(x, y_neg, 'b-')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Parabola: y^2 = {k}x')
plt.grid(True)
plt.legend()
plt.show()
```
