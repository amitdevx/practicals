### 1.

Represent the following information using a pie chart , by using explode function to separate sectors.

| Subject | Marks |
| --- | --- |
| Computer | 60 |
| Mathematics | 80 |
| Electronics | 65 |
| Statistics | 55 |
| English | 70 |



```python
import matplotlib.pyplot as plt

subjects = ['Computer', 'Mathematics', 'Electronics', 'Statistics', 'English']
marks = [60, 80, 65, 55, 70]
explode = [0.05] * len(subjects)

fig, ax = plt.subplots(figsize=(6.5, 6.5))
ax.pie(marks, labels=subjects, autopct='%1.1f%%', explode=explode,
       startangle=90, shadow=True)
ax.set_title('Marks Distribution by Subject')
ax.axis('equal')
plt.show()
```

### 2.

Plot the graph of f(x) = sin(3 x + 1) , for x  [–5 π, 5 π] in red colour and dotted pattern.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5 * np.pi, 5 * np.pi, 1200)
y = np.sin(3 * x + 1)

plt.figure(figsize=(9, 4))
plt.plot(x, y, 'r:', linewidth=2, label=r'$f(x) = \sin(3x + 1)$')
plt.title(r'Plot of $f(x) = \sin(3x + 1)$ on $[-5\pi, 5\pi]$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True, linestyle=':')
plt.show()
```

### 3.

Draw surface plot for the function f(x, y) = – ( x2 +y2 ) for x, y  [–10, 20].


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

x = np.linspace(-10, 20, 150)
y = np.linspace(-10, 20, 150)
X, Y = np.meshgrid(x, y)
Z = -(X**2 + Y**2)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title(r'Surface Plot of $f(x, y) = -(x^2 + y^2)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```

### 4.

Using sympy find the point of intersection of two lines L1 and L2, given by the equations
x + y = 2 and 3x+2y =1. Also find their slopes.


```python
from sympy import symbols, Eq, solve, Line

x, y = symbols('x y')
L1 = Line((0, 2), (-2, 0))  # x + y = 2
L2 = Line((0, 0.5), (-1/3, 0))  # 3x + 2y = 1

intersection = L1.intersection(L2)[0]
print(f'Intersection point of L1 and L2: {intersection}')
print(f'Slope of L1: {L1.slope}')
print(f'Slope of L2: {L2.slope}')
```

### 5.

Using sympy declare a regular polygon with 6 sides and radius 1 centered at (1, 2) and find its
area perimeter and centroid.


```python
from sympy import Point, RegularPolygon

center = Point(1, 2)
polygon = RegularPolygon(center, 1, 6)

print(f'Area of the regular hexagon: {polygon.area}')
print(f'Perimeter of the regular hexagon: {polygon.perimeter}')
print(f'Centroid of the regular hexagon: {polygon.centroid}')
```

1. Using python program find the combined transformation of the line segment joining
the
points P[5,−2] and Q[3, 2] for the following sequence of transformations:-
(a) Rotation about origin through an angle π. (b) Scaling in X−coordinate by 4 units.
(c) Reflection through the line y = x. Also plot original and transformed line segments.

```python
import numpy as np
import matplotlib.pyplot as plt

P = np.array([5.0, -2.0])
Q = np.array([3.0, 2.0])

rotation = np.array([[-1, 0], [0, -1]], dtype=float)  # rotation by pi
scale_x = np.array([[4, 0], [0, 1]], dtype=float)
reflect_yx = np.array([[0, 1], [1, 0]], dtype=float)  # reflection about y = x

transform_matrix = reflect_yx @ scale_x @ rotation

P_trans = transform_matrix @ P
Q_trans = transform_matrix @ Q

plt.figure(figsize=(6, 6))
plt.plot([P[0], Q[0]], [P[1], Q[1]], color='blue', marker='o', label='Original PQ')
plt.plot([P_trans[0], Q_trans[0]], [P_trans[1], Q_trans[1]], color='red', marker='s', label='Transformed PQ')

plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.title('Combined Transformations of Line Segment PQ')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle=':')
plt.show()
```

2. Write python program to generate and plot100 points on circle x2 +y2 = 49 using iterative
formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

radius = 7
num_points = 100
theta = 0.0
step = 2 * np.pi / num_points

xs = []
ys = []
for _ in range(num_points):
    xs.append(radius * np.cos(theta))
    ys.append(radius * np.sin(theta))
    theta += step

plt.figure(figsize=(6, 6))
plt.plot(xs, ys, marker='o', linestyle='-', color='teal', label=r'$x^2 + y^2 = 49$')
plt.gca().set_aspect('equal', 'box')
plt.title('Points on Circle $x^2 + y^2 = 49$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```
