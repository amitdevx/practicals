### 3.

Draw surface plot for the function 𝑓(𝑥, 𝑦) = √𝑥 2 − 𝑦 2 when –30 x, y 30,
with brown colour.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

grid = np.linspace(-30, 30, 200)
X, Y = np.meshgrid(grid, grid)
Z = np.sqrt(np.clip(X**2 - Y**2, a_min=0, a_max=None))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, color="saddlebrown", edgecolor="none", alpha=0.9)
ax.set_title(r"Surface Plot of $f(x, y) = \sqrt{x^2 - y^2}$")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
plt.show()
```

### 4.

Using sympy declare line AB passing through points A(2, –4), B(2,8) and line PQ passing
through point P(5,4), Q(–8,3). Find point of intersection of lines AB and PQ, also equations of
Lines AB and PQ, and distance between point A and line PQ.


```python
from sympy import Point, Line

A = Point(2, -4)
B = Point(2, 8)
P = Point(5, 4)
Q = Point(-8, 3)

line_ab = Line(A, B)
line_pq = Line(P, Q)

intersection = line_ab.intersection(line_pq)
distance_a_to_pq = line_pq.distance(A)

print(f"Equation of line AB: {line_ab.equation()}")
print(f"Equation of line PQ: {line_pq.equation()}")
print(f"Intersection of AB and PQ: {intersection[0] if intersection else 'No intersection'}")
print(f"Distance from point A to line PQ: {distance_a_to_pq.evalf()}")
```

### 5.

Using sympy declare triangle OAB where O (0, 0), A(7, –2), B( 3, 6) , for this triangle find the
scaling in x and y co-ordinates by factors 3 and –7 respectively and also find its reflection
through the line y = x.


```python
from sympy import Matrix

O = Matrix([0, 0])
A = Matrix([7, -2])
B = Matrix([3, 6])

scaling_matrix = Matrix([[3, 0], [0, -7]])
scaled_O = scaling_matrix * O
scaled_A = scaling_matrix * A
scaled_B = scaling_matrix * B

reflection_matrix = Matrix([[0, 1], [1, 0]])  # reflection about y = x
reflected_O = reflection_matrix * O
reflected_A = reflection_matrix * A
reflected_B = reflection_matrix * B

print(f"Scaled vertices (sx=3, sy=-7): O{tuple(scaled_O)}, A{tuple(scaled_A)}, B{tuple(scaled_B)}")
print(f"Reflected vertices about y = x: O{tuple(reflected_O)}, A{tuple(reflected_A)}, B{tuple(reflected_B)}")
```

1. Write a python program to plot the 3d line segment with vertices at A(1, 2, 3), B(4 , 5 , 6).
Find and plot a) reflection of AB through XY-plane, b) shearing of AB in y co-ordinate
proportional to z co-ordinate by factor 4.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

def segment_endpoints(start, end):
    return np.vstack([start, end])

original = segment_endpoints(A, B)
reflection_xy = segment_endpoints(A * np.array([1, 1, -1]), B * np.array([1, 1, -1]))
shear_factor = 4
shear_matrix = np.array([[1, 0, 0], [0, 1, shear_factor], [0, 0, 1]])
sheared = segment_endpoints(shear_matrix @ A, shear_matrix @ B)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.plot(original[:, 0], original[:, 1], original[:, 2], color="blue", label="Original AB", marker="o")
ax.plot(
    reflection_xy[:, 0],
    reflection_xy[:, 1],
    reflection_xy[:, 2],
    color="green",
    label="Reflection through XY-plane",
    marker="o",
)
ax.plot(
    sheared[:, 0],
    sheared[:, 1],
    sheared[:, 2],
    color="red",
    label="Shear in y (factor 4·z)",
    marker="o",
)

ax.set_title("Transformations of Line Segment AB")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()
plt.show()
```

2. Write a python program to generate and plot 50 points on the parabola with equation y2 = 8x
where 0 ≤ 𝑥 ≤ 30 using iterative formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

num_points = 50
x_values = np.linspace(0, 30, num_points // 2)
points = []
for x in x_values:
    y = np.sqrt(8 * x)
    points.append((x, y))
    points.append((x, -y))

points = np.array(points)

plt.figure(figsize=(7, 5))
plt.scatter(points[:, 0], points[:, 1], color="darkorange", label=r"$y^2 = 8x$")
plt.title("Points on Parabola via Iterative Formula")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0, 30)
plt.grid(True, linestyle=":")
plt.legend()
plt.show()
```
