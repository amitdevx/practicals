Write python program to plot function 𝑓(𝑥) = 𝑥 2 for 𝑥 ∈ [−10,10] and 𝑔(𝑥) = 𝑒 2𝑥+4
for 𝑥 ∈ [1,20] in one graph window by using 1 × 2 subplots.


```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-10, 10, 400)
f_x = x1**2

x2 = np.linspace(1, 20, 400)
g_x = np.exp(2 * x2 + 4)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(x1, f_x, color="navy", linewidth=2)
axes[0].set_title(r"$f(x) = x^2$")
axes[0].set_xlabel("x")
axes[0].set_ylabel("f(x)")
axes[0].grid(True, linestyle=":")

axes[1].plot(x2, g_x, color="darkred", linewidth=2)
axes[1].set_title(r"$g(x) = e^{2x+4}$")
axes[1].set_xlabel("x")
axes[1].set_ylabel("g(x)")
axes[1].grid(True, linestyle=":")

plt.tight_layout()
plt.show()
```

### 3.

Draw Wireframes plot for the function f(x, y) = sin(x2+y2) when 0 x, y 2π,
with brown colour.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

grid = np.linspace(0, 2 * np.pi, 120)
X, Y = np.meshgrid(grid, grid)
Z = np.sin(X**2 + Y**2)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.plot_wireframe(X, Y, Z, color="saddlebrown", linewidth=0.7)
ax.set_title(r"Wireframe of $f(x, y) = \sin(x^2 + y^2)$")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
plt.show()
```

### 4.

Using sympy declare the regular polygon with center at origin, radius 4 and number of sides 5.
Find its area and also find its scaling in x and y directions by factors 3 and 2 respectively.


```python
from sympy import Point, RegularPolygon

polygon = RegularPolygon(Point(0, 0), 4, 5)
area = polygon.area
scaled_vertices = [Point(3 * v.x, 2 * v.y) for v in polygon.vertices]

print(f"Regular polygon vertices: {polygon.vertices}")
print(f"Area: {area}")
print("Scaled vertices with factors (sx=3, sy=2):")
for vertex in scaled_vertices:
    print(vertex)
```

### 5.

Using sympy declare polygon ABCD with vertices A(2, 2), B(6, 2), C(6, 6) and D(2, 6). Find
its area, perimeter and all angles.


```python
from sympy import Point, Polygon, pi

A = Point(2, 2)
B = Point(6, 2)
C = Point(6, 6)
D = Point(2, 6)
polygon = Polygon(A, B, C, D)

area = polygon.area
perimeter = polygon.perimeter
angles = {vertex: float(angle.evalf() * 180 / pi) for vertex, angle in polygon.angles.items()}

print(f"Polygon vertices: {polygon.vertices}")
print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
for vertex, angle in angles.items():
    print(f"Angle at {vertex}: {angle:.2f} degrees")
```

### Q2.1 (Attempt Any One)

1. Write a python program to plot the  ABC with vertices A(1, 6), B(0 , 8) and C(5 , 3). Find
and its 1) shearing in x-direction proportional to y by factor 5, 2) reflection through y-axis ,
3) translation x and y direction by factor 1 and 4 units respectively. And plot all transformed
figures in different colours.

```python
import numpy as np
import matplotlib.pyplot as plt

vertices = np.array([[1, 6], [0, 8], [5, 3]])

def close_shape(points):
    return np.vstack([points, points[0]])

original = close_shape(vertices)

shear_factor = 5
shear_matrix = np.array([[1, shear_factor], [0, 1]])
sheared = close_shape((shear_matrix @ vertices.T).T)

reflection_matrix = np.array([[-1, 0], [0, 1]])
reflected = close_shape((reflection_matrix @ sheared[:-1].T).T)

translation = np.array([1, 4])
translated_vertices = reflected[:-1] + translation
translated = close_shape(translated_vertices)

fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(original[:, 0], original[:, 1], label="Original", linewidth=2)
ax.plot(sheared[:, 0], sheared[:, 1], label="Sheared (x by 5·y)", linewidth=2)
ax.plot(reflected[:, 0], reflected[:, 1], label="Reflected (y-axis)", linewidth=2)
ax.plot(translated[:, 0], translated[:, 1], label="Translated (+1, +4)", linewidth=2)

ax.set_aspect("equal")
ax.set_title("Transformations on Triangle ABC")
ax.legend()
ax.grid(True, linestyle=":")
plt.show()
```

2. Write a Python program to generate and plot 75 points on circle x2+y2 = 36 using iterative
formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

radius = 6
num_points = 75
theta = 2 * np.pi / num_points
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

points = []
vector = np.array([radius, 0.0])
for _ in range(num_points):
    points.append(vector.copy())
    vector = rotation_matrix @ vector

points = np.array(points)

plt.figure(figsize=(6, 6))
plt.scatter(points[:, 0], points[:, 1], color="slateblue", s=25, label="Circle points")
plt.gca().add_patch(plt.Circle((0, 0), radius, fill=False, linestyle="--", color="gray"))
plt.title(r"Iterative Points on $x^2 + y^2 = 36$")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.grid(True, linestyle=":")
plt.legend()
plt.show()
```
