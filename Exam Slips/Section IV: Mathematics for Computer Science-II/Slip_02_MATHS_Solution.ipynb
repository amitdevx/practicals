### 1.

Represent the following information using a Pie chart with percentage.

| Sport | Number of students |
| --- | --- |
| Cricket | 34 |
| Football | 50 |
| Badminton | 24 |
| Hockey | 10 |
| Other | 82 |



```python
import matplotlib.pyplot as plt

sports = ["Cricket", "Football", "Badminton", "Hockey", "Other"]
students = [34, 50, 24, 10, 82]
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99", "#c2c2f0"]

plt.figure(figsize=(7, 7))
plt.pie(
    students,
    labels=sports,
    autopct="%1.1f%%",
    startangle=90,
    colors=colors,
    shadow=True,
)
plt.title("Students Participating in Sports")
plt.axis("equal")
plt.show()
```

### 2.

Plot the graph of 𝑦 = 𝑥 2 + 5, for x  [–5, 5] in green colour with dashed lines.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 300)
y = x**2 + 5

plt.figure(figsize=(8, 5))
plt.plot(x, y, color="green", linestyle="--", linewidth=2)
plt.title(r"Graph of $y = x^2 + 5$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, linestyle=":", alpha=0.5)
plt.show()
```

### 3.

Draw Wireframes plot for the function f(x, y) = (x2 – y2) when –20 x, y 20,
with brown colour.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

grid = np.linspace(-20, 20, 150)
X, Y = np.meshgrid(grid, grid)
Z = X**2 - Y**2

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.plot_wireframe(X, Y, Z, color="saddlebrown", linewidth=0.7)
ax.set_title(r"Wireframe of $f(x, y) = x^2 - y^2$")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
plt.show()
```

### 4.

Using sympy declare point P(–4,5), apply each of the given transformation on point P.
a) Scaling in x co-ordinate by 2 unit, b) reflection through the line y = x, c) translation in y
direction by factor 7 units.


```python
from sympy import Point

P = Point(-4, 5)
scaled = Point(2 * P.x, P.y)
reflected = Point(P.y, P.x)  # reflection about y = x swaps coordinates
translated = Point(P.x, P.y + 7)

print(f"Original point P: {P}")
print(f"After scaling x by 2 units: {scaled}")
print(f"After reflection about y = x: {reflected}")
print(f"After translation +7 units along y: {translated}")
```

### 5.

Using sympy declare  ABC where A (3 , 0), B(0, 4), C(–3, 0) , for this triangle find its area,
perimeter and all angles.


```python
from sympy import Point, Polygon, pi

A = Point(3, 0)
B = Point(0, 4)
C = Point(-3, 0)
triangle = Polygon(A, B, C)

area = triangle.area
perimeter = triangle.perimeter
angles_deg = {vertex: float(angle.evalf() * 180 / pi) for vertex, angle in triangle.angles.items()}

print(f"Triangle vertices: {triangle.vertices}")
print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
for vertex, angle in angles_deg.items():
    print(f"Angle at {vertex}: {angle:.2f} degrees")
```

### Q2.1 (Attempt Any One)

1. Write a python program to plot the polygon with vertices A(1, 1), B(–5, 1), C(–5, – 4) and
D(1, – 4). On this polygon apply 1) rotation about origin through angle 900 , 2) shearing in
x-direction proportional to y by factor 3.
Plot transformed polygons also. All figures must be in different colours. Use legend.

```python
import numpy as np
import matplotlib.pyplot as plt

vertices = np.array([[1, 1], [-5, 1], [-5, -4], [1, -4]])

def close_shape(points):
    return np.vstack([points, points[0]])

original = close_shape(vertices)

rotation_angle = np.deg2rad(90)
rotation_matrix = np.array(
    [[np.cos(rotation_angle), -np.sin(rotation_angle)], [np.sin(rotation_angle), np.cos(rotation_angle)]]
)
rotated = close_shape((rotation_matrix @ vertices.T).T)

shear_factor = 3
shear_matrix = np.array([[1, shear_factor], [0, 1]])
rotated_sheared = close_shape((shear_matrix @ rotated[:-1].T).T)

fig, ax = plt.subplots(figsize=(7, 7))
ax.plot(original[:, 0], original[:, 1], color="blue", linewidth=2, label="Original")
ax.plot(rotated[:, 0], rotated[:, 1], color="green", linewidth=2, label="Rotated 90°")
ax.plot(rotated_sheared[:, 0], rotated_sheared[:, 1], color="red", linewidth=2, label="Rotation + X-shear")

ax.set_aspect("equal")
ax.set_title("Polygon Transformations")
ax.legend()
ax.grid(True, linestyle=":")
plt.show()
```

2. Write a python program to generate and plot 100 points on circle x2+y2 = 49 using iterative
formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

radius = 7
num_points = 100
theta = 2 * np.pi / num_points
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

points = []
vector = np.array([radius, 0.0])
for _ in range(num_points):
    points.append(vector.copy())
    vector = rotation_matrix @ vector

points = np.array(points)

plt.figure(figsize=(6, 6))
plt.scatter(points[:, 0], points[:, 1], color="teal", s=25, label="Generated points")
plt.gca().add_patch(plt.Circle((0, 0), radius, fill=False, linestyle="--", color="gray"))
plt.title(r"Iterative Points on $x^2 + y^2 = 49$")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.grid(True, linestyle=":")
plt.legend()
plt.show()
```
