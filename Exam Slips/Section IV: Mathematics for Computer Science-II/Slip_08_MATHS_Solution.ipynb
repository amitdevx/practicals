### 1.

Represent the following information using a bar graph .

| City | Temperature in ℃ |
| --- | --- |
| Satara | 20 |
| Dhule | 32 |
| Nashik | 25 |
| Nagpur | 40 |
| Pune | 30 |



```python
import matplotlib.pyplot as plt
cities = ['Satara', 'Dhule', 'Nashik', 'Nagpur', 'Pune']

temps = [20, 32, 25, 40, 30]

plt.figure()

plt.bar(cities, temps, color='orange')

plt.xlabel('City')

plt.ylabel('Temperature (C)')

plt.title('City Temperatures')

plt.show()
```

### 2.

Plot the graph of 𝑦 = cos(𝑥) + sin (𝑥), for x  [–𝜋, 𝜋] in maroon colour with dashed lines.


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-np.pi, np.pi, 100)

y = np.cos(x) + np.sin(x)

plt.figure()

plt.plot(x, y, color='maroon', linestyle='--', label='y = cos(x) + sin(x)')

plt.xlabel('x')

plt.ylabel('y')

plt.legend()

plt.grid(True)

plt.show()
```

### 3.

Draw Wireframe plot for the function 𝑓(𝑥, 𝑦) = sin (√𝑥 2 + 𝑦 2 ) when 0 x, y 2𝜋,
with brown colour.


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2*np.pi, 50)

y = np.linspace(0, 2*np.pi, 50)

X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(X, Y, Z, color='brown')

ax.set_title('Wireframe Sin(r)')

plt.show()
```

### 4.

Using sympy declare points L(2, 6) , M(–2, 9) and N(–3, –5). Determine whether the points L,
M and N are collinear, find distance between L and M, distance between N and M and distance
between L and N and slope of line passing through Point L and M.


```python
import sympy as sp
L = sp.Point(2, 6)

M = sp.Point(-2, 9)

N = sp.Point(-3, -5)

collinear = sp.Point.is_collinear(L, M, N)

print(f"L, M, N Collinear: {collinear}")

d_LM = L.distance(M)

d_MN = M.distance(N)

d_LN = L.distance(N)

print(f"Distance LM: {d_LM}")

print(f"Distance MN: {d_MN}")

print(f"Distance LN: {d_LN}")

slope_LM = sp.Line(L, M).slope

print(f"Slope LM: {slope_LM}")
```

### 5.

Using sympy declare regular polygon with center at origin, radius 6 and number of sides 6. Find
its area, perimeter and translate this regular polygon in x direction by 8 units.


```python
import sympy as sp
poly = sp.RegularPolygon(sp.Point(0, 0), 6, 6)

print(f"Area: {poly.area}")

print(f"Perimeter: {poly.perimeter}")

new_verts = [p.translate(8, 0) for p in poly.vertices]

print("Translated Vertices (x+8):", new_verts)
```

1. Write a python program to plot the polygon with vertices A(–2 , 5), B(–2, 9) , C(3 , 5) and
D(4, 2) . Apply following transformations on polygon ABCD and plot the transformed
polygons also, a) translation in x and y direction by factor 2 and –5 respectively, b) rotation
about origin through angle 600 and c) shearing in x direction by factor 5 units.

```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
def apply_transformation(points, matrix):
    return np.dot(points, matrix.T)

poly = np.array([[-2, 5], [-2, 9], [3, 5], [4, 2]])

poly_plot = np.vstack([poly, poly[0]])

trans = np.array([2, -5])

theta = np.radians(60)

c, s = np.cos(theta), np.sin(theta)

R = np.array([[c, -s], [s, c]])

Sh = np.array([[1, 5], [0, 1]])

p1 = poly + trans

p1_plot = np.vstack([p1, p1[0]])

p2 = apply_transformation(p1, R)

p2_plot = np.vstack([p2, p2[0]])

p3 = apply_transformation(p2, Sh)

p3_plot = np.vstack([p3, p3[0]])

plt.figure()

plt.plot(poly_plot[:, 0], poly_plot[:, 1], 'b-', label='Original')

plt.plot(p1_plot[:, 0], p1_plot[:, 1], 'g--', label='Translated')

plt.plot(p2_plot[:, 0], p2_plot[:, 1], 'c--', label='Rotated')

plt.plot(p3_plot[:, 0], p3_plot[:, 1], 'r-', label='Sheared')

plt.legend()

plt.grid(True)

plt.axis('equal')

plt.title("Polygon Transformations")

plt.show()
```

2. Write a python program to generate and plot 100 points on circle 𝑥 2 + 𝑦 2 = 81 using
iterative formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

r = 9.0
n_points = 100

theta = np.linspace(0, 2*np.pi, n_points)

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure()
plt.plot(x, y, label=f'x^2 + y^2 = 81')
plt.title(f'Circle x^2 + y^2 = 81 (Iterative)')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal') # Ensure circle looks circular
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
