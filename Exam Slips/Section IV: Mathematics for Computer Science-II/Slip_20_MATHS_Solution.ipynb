| Subject | Number of students |

```python
import matplotlib.pyplot as plt
from sympy import Point

A = Point(1, 1)
B = Point(1, 6)
C = Point(6, 1)
D = Point(6, 6)

points = [A, B, D, C, A]
xs = [pt.x for pt in points]
ys = [pt.y for pt in points]

plt.figure(figsize=(6, 6))
plt.plot(xs, ys, color='black', marker='o')
plt.title('Quadrilateral ABCD')
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal', 'box')
plt.grid(True, linestyle=':')
plt.show()
```

### 4.

Using sympy declare quadrilateral ABCD where A (1 ,1), B(1, 4), C( 4, 1) and D(4, 4),
and find shearing of ABCD in x and y directions by 3 and 7 units respectively


```python
from sympy import Point, Matrix

A = Point(1, 1)
B = Point(1, 4)
C = Point(4, 1)
D = Point(4, 4)
points = [A, B, C, D]

shear_x = Matrix([[1, 3], [0, 1]])
shear_y = Matrix([[1, 0], [7, 1]])

def apply(matrix, point):
    vec = matrix * Matrix(point)
    return Point(vec[0], vec[1])

sheared_x = [apply(shear_x, pt) for pt in points]
sheared_y = [apply(shear_y, pt) for pt in points]

print('Shear in x-direction (k = 3):')
for label, pt in zip(["A'", "B'", "C'", "D'"], sheared_x):
    print(f'  {label}: {pt}')

print('Shear in y-direction (k = 7):')
for label, pt in zip(['A"', 'B"', 'C"', 'D"'], sheared_y):
    print(f'  {label}: {pt}')
```

If the line segment joining the points A[−3, 5] and B[3, 3] is transformed to the line segment

```python
from sympy import Point, Matrix

A = Point(-3, 5)
B = Point(3, 3)
T = Matrix([[1/4, 3], [1, 1]])

def transform(point):
    vec = T * Matrix(point)
    return Point(vec[0], vec[1])

A_prime = transform(A)
B_prime = transform(B)
midpoint = Point((A_prime.x + B_prime.x) / 2, (A_prime.y + B_prime.y) / 2)

print(f"A' transformed: {A_prime}")
print(f"B' transformed: {B_prime}")
print(f"Midpoint of A'B': {midpoint}")
```

1. Write a python program to plot the triangle with vertices at (4,4) , (2, 4) ,( 4, 2) .
𝜋
Apply the transformations a) its rotation about the origin by radians, b) uniform scaling
2
by factor 3. Plot original and transformed figures in different colours
Also plot x and y
axes in one colour.

```python
import numpy as np
import matplotlib.pyplot as plt

triangle = np.array([[4, 4], [2, 4], [4, 2]], dtype=float)
closed_original = np.vstack([triangle, triangle[0]])

theta = np.pi / 2
rotation = np.array([[np.cos(theta), -np.sin(theta)],
                     [np.sin(theta), np.cos(theta)]])
rotated = (rotation @ triangle.T).T
closed_rotated = np.vstack([rotated, rotated[0]])

scaled = triangle * 3
closed_scaled = np.vstack([scaled, scaled[0]])

plt.figure(figsize=(6, 6))
plt.plot(closed_original[:, 0], closed_original[:, 1], color='blue', marker='o', label='Original Triangle')
plt.plot(closed_rotated[:, 0], closed_rotated[:, 1], color='red', marker='s', label=r'Rotated $90^\circ$')
plt.plot(closed_scaled[:, 0], closed_scaled[:, 1], color='green', marker='^', label='Uniform Scale x3')

plt.axhline(0, color='black', linewidth=0.8, label='Axes')
plt.axvline(0, color='black', linewidth=0.8)
plt.gca().set_aspect('equal', 'box')
plt.title('Transformations of Triangle')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle=':')
plt.show()
```

2. Write Python program to generate and plot 100 points on parabola y2 = 32 x ,
iterative formulae for 0  x  100.

```python
import matplotlib.pyplot as plt

num_points = 100
max_x = 100
step = max_x / (num_points - 1)

xs = []
ys = []
x_val = 0.0
for _ in range(num_points):
    y_val = (32 * x_val) ** 0.5
    xs.append(x_val)
    ys.append(y_val)
    x_val += step

plt.figure(figsize=(7, 4))
plt.scatter(xs, ys, color='purple', label=r'$y^2 = 32x$')
plt.title(r'Points on Parabola $y^2 = 32x$ for $0 \leq x \leq 100$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```
