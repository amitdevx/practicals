### 1.

Represent the following information using a Pie Chart showing percentage.

| Subject | Marks |
| --- | --- |
| Mathematics | 85 |
| English | 15 |
| Biology | 50 |
| Physics | 67 |

```python
import matplotlib.pyplot as plt
subjects = ['Mathematics', 'English', 'Biology', 'Physics', 'Chemistry']

marks = [85, 15, 50, 67, 77]

plt.figure()

plt.pie(marks, labels=subjects, autopct='%1.1f%%')

plt.title('Subject Marks Percentage')

plt.show()
```

### 2.

Plot the graph of y = cos ( x2) , for x  [ -π, π] in red colour with dashed lines.


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-np.pi, np.pi, 100)

y = np.cos(x**2)

plt.figure()

plt.plot(x, y, 'r--', label='y = cos(x^2)')

plt.xlabel('x')

plt.ylabel('y')

plt.legend()

plt.grid(True)

plt.show()
```

### 3.

Draw wireframe plot for the function f (x, y) = (x2 – y2) when –10  x, y  10,
with brown colour.


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 30)

y = np.linspace(-10, 10, 30)

X, Y = np.meshgrid(x, y)

Z = X**2 - Y**2

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(X, Y, Z, color='brown')

ax.set_title('Wireframe Plot')

plt.show()
```

### 4.

Using sympy declare points P ( 0, 0 ) , Q (1 , 1 ), R (–1 ,1) , S (–1, –1 ), T ( 1 , –1 ) , Rotate the
polygon about origin through π radians also find centroid of polygon.


```python
import sympy as sp
poly = sp.Polygon(sp.Point(0,0), sp.Point(1,1), sp.Point(-1,1), sp.Point(-1,-1), sp.Point(1,-1))

print(f"Centroid: {poly.centroid}")

rotated_verts = [p.rotate(sp.pi, sp.Point(0,0)) for p in poly.vertices]

print("Rotated Vertices (pi):", rotated_verts)
```

### Q2.1 (Attempt Any One)

1. Write a python program to plot the Polygon ABCD where A (0 , 3), B (–3, 3 ), C ( –3, 0 ) and
D ( 0 ,0). Find and plot, 1) reflection of ABCD through line y = x , 2) Uniform Scaling of
ABCD by facor 4. All the figures must be in different colours.

```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
def apply_transformation(points, matrix):
    return np.dot(points, matrix.T)

poly = np.array([[0, 3], [-3, 3], [-3, 0], [0, 0]])

poly_plot = np.vstack([poly, poly[0]])

Ref = np.array([[0, 1], [1, 0]])

Scale = np.diag([4, 4])

p_ref = apply_transformation(poly, Ref)

p_ref_plot = np.vstack([p_ref, p_ref[0]])

p_scale = apply_transformation(poly, Scale)

p_scale_plot = np.vstack([p_scale, p_scale[0]])

plt.figure()

plt.plot(poly_plot[:, 0], poly_plot[:, 1], 'b-', label='Original')

plt.plot(p_ref_plot[:, 0], p_ref_plot[:, 1], 'g--', label='Refl y=x')

plt.plot(p_scale_plot[:, 0], p_scale_plot[:, 1], 'r-.', label='Scale 4')

plt.legend()

plt.grid(True)

plt.axis('equal')

plt.title("Polygon Transformations")

plt.show()
```

### Q2.2

2. Write a Python program to generate and plot 50 points on circle x2 +y2 = 25 using iterative
formulae.

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

# Parameters
r_sq = 25
r = np.sqrt(r_sq)
num_points = 50

# Generate points using iterative formula (parametric approach)
theta = np.linspace(0, 2*np.pi, num_points)
x = r * np.cos(theta)
y = r * np.sin(theta)

# Plot
plt.figure(figsize=(6,6))
plt.plot(x, y, label=f'x^2 + y^2 = {r_sq}')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Circle: x^2 + y^2 = {r_sq}')
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```
