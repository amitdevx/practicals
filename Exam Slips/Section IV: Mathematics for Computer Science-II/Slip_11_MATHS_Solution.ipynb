### 1.

Draw the Histograph with given data in brown colour,which will give us the number of
students obtaining the marks between 20– 40 ,40– 60 , 60 – 80 and 80–100

### 2.

Plot the graph of 𝑦 = 𝑥 2 + cos (2𝑥 + 5), for x  [0, 2 ] in red colour with arrow lines.


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, np.pi/2, 50)

y = x**2 + np.cos(2*x + 5)

plt.figure()

plt.plot(x, y, 'r-', marker='>', label='y = x^2 + cos(2x+5)')

plt.xlabel('x')

plt.ylabel('y')

plt.legend()

plt.grid(True)

plt.show()
```

### 3.

Draw Surface plot for the function 𝑓(𝑥, 𝑦) = 𝑆𝑖𝑛 √𝑥 2 + 𝑦 2 when 0 x, y 2𝜋,
with brown colour.


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2*np.pi, 50)

y = np.linspace(0, 2*np.pi, 50)

X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(X, Y, Z, color='brown')

ax.set_title('Wireframe Sin(r)')

plt.show()
```

### 4.

Using sympy generate the lines with equations 3 x+ 5y =4 and x – 5y =8. Find point of
intersection of these two lines and angle between these two lines.


```python
import numpy as np
import sympy as sp
x, y = sp.symbols('x y')

L1 = sp.Line(3*x + 5*y - 4)

L2 = sp.Line(x - 5*y - 8)

intersection = L1.intersection(L2)

angle = L1.angle_between(L2)

print(f"Intersection: {intersection}")

print(f"Angle (radians): {angle}")

print(f"Angle (degrees): {np.degrees(float(angle))}")
```

### 5.

Using sympy declare polygon with vertices O(0, 0), A(5, 0), C(5, 2) and D(0, 2). Determine
whether point (5,1) is inside the polygon. Also find the translation of polygon in x and y
direction by factor 3 and 3 units respectively.


```python
import sympy as sp
poly = sp.Polygon(sp.Point(0, 0), sp.Point(5, 0), sp.Point(5, 2), sp.Point(0, 2))

P = sp.Point(5, 1)

is_inside = poly.encloses_point(P)

on_boundary = False

for side in poly.sides:
    if side.contains(P):
        on_boundary = True
        break

print(f"Point (5,1) is strictly inside: {is_inside}")

print(f"Point (5,1) is on boundary: {on_boundary}")

new_verts = [p.translate(3, 3) for p in poly.vertices]

print("Translated Vertices:", new_verts)
```

1. Write a python program to generate and plot the  ABC, where A(10, 5), B(–7, 4) and
C(8, –3). Find and plot the transformed triangle by applying combined transformations of
sequence a) reflection through y-axis, b) rotation about origin through angle 900 and c)
uniform scaling by factor 7 units.

```python
import matplotlib.pyplot as plt
import numpy as np
def apply_transformation(points, matrix):
    return np.dot(points, matrix.T)

poly = np.array([[10, 5], [-7, 4], [8, -3]])

poly_plot = np.vstack([poly, poly[0]])

Ref = np.diag([-1, 1])

theta = np.radians(90)

c, s = np.cos(theta), np.sin(theta)

Rot = np.array([[0, -1], [1, 0]])

Scale = np.diag([7, 7])

p1 = apply_transformation(poly, Ref)

p2 = apply_transformation(p1, Rot)

p3 = apply_transformation(p2, Scale)

p3_plot = np.vstack([p3, p3[0]])

plt.figure()

plt.plot(poly_plot[:, 0], poly_plot[:, 1], 'b-', label='Original')

plt.plot(p3_plot[:, 0], p3_plot[:, 1], 'r-', label='Final Transformed')

plt.legend()

plt.grid(True)

plt.axis('equal')

plt.title("Combined Transformations")

plt.show()
```

2. Write a Python program to generate and plot 50 points on parabola 𝑦 2 = 40𝑥 for 0 ≤ 𝑥 ≤ 50
using iterative formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

a = 10.0
x_limit = 50
n_points = 50

t_max = np.sqrt(x_limit / a)
t = np.linspace(-t_max, t_max, n_points)

x = a * t**2
y = 2 * a * t

plt.figure()
plt.plot(x, y, label=f'y^2 = 40x')
plt.title(f'Parabola y^2 = 40x (Iterative)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
