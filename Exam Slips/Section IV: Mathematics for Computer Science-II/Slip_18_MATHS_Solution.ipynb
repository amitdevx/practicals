### 1.

The age of 25 persons who donated blood in camp arranged for the age group 20 to 45 yrs is

### 3.

Draw surface plot for the function f(x, y) = (–x2 –y2) for x, y  [–15, 10], in Green
colour.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

x = np.linspace(-15, 10, 150)
y = np.linspace(-15, 10, 150)
X, Y = np.meshgrid(x, y)
Z = -(X**2 + Y**2)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='green', edgecolor='none', alpha=0.9)
ax.set_title(r'Surface Plot of $f(x, y) = -(x^2 + y^2)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```

### 4.

Using sympy declare two lines L1 and L2 given by the equations x + y =2 and 3x+2y =1. Find
the point of intersection of lines L1 and L2.

Also find the angle between these two lines.


```python
from sympy import Line

L1 = Line((0, 2), (-2, 0))  # x + y = 2
L2 = Line((0, 0.5), (-1/3, 0))  # 3x + 2y = 1

intersection = L1.intersection(L2)[0]
angle = L1.angle_between(L2)

print(f'Intersection point: {intersection}')
print(f'Angle between L1 and L2 (radians): {angle}')
```

### 5.

Using sympy declare polygon P with vertices A(1 , 3), B( 3, 5), C(5 , 4), D ( 4, 2), Determine
𝜋
whether it is convex polygon? Rotate polygon P about point (2,3) by angle .
3


```python
from sympy import Point, Polygon
from sympy import pi

A = Point(1, 3)
B = Point(3, 5)
C = Point(5, 4)
D = Point(4, 2)
polygon = Polygon(A, B, C, D)

print(f'Is polygon ABCD convex? {polygon.is_convex()}')

center = Point(2, 3)
rotated_points = [pt.rotate(pi / 3, center) for pt in (A, B, C, D)]
for label, pt in zip(["A'", "B'", "C'", "D'"], rotated_points):
    print(f'Rotated point {label}: {pt}')
```

### Q2.1 (Attempt Any One)

1. Write a python program to generate and plot the  PQR with vertices at P (1,
2), Q(4 2),
R(4, 5). And also plot its uniform scaling by factor 0.5 and uniform scaling by factor 4.
All three figures must be in different colours.

```python
import numpy as np
import matplotlib.pyplot as plt

triangle = np.array([[1, 2], [4, 2], [4, 5]], dtype=float)

scales = [0.5, 1.0, 4.0]
colors = ['red', 'blue', 'green']
labels = ['Scale 0.5', 'Original', 'Scale 4']

plt.figure(figsize=(6, 6))
for scale, color, label in zip(scales, colors, labels):
    scaled = triangle * scale
    closed = np.vstack([scaled, scaled[0]])
    plt.plot(closed[:, 0], closed[:, 1], marker='o', color=color, label=label)

plt.title('Uniform Scaling of Triangle PQR')
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal', 'box')
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```

2. Write python program to generate and plot 60 on the circle with equation x2 +y2 = 100 using
iterative formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

radius = 10
num_points = 60
theta = 0.0
step = 2 * np.pi / num_points

xs = []
ys = []
for _ in range(num_points):
    xs.append(radius * np.cos(theta))
    ys.append(radius * np.sin(theta))
    theta += step

plt.figure(figsize=(6, 6))
plt.plot(xs, ys, color='orange', marker='o', label=r'$x^2 + y^2 = 100$')
plt.gca().set_aspect('equal', 'box')
plt.title('Points on Circle $x^2 + y^2 = 100$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```
