### 1.

Show the following information by pie graph. Use labels and display percentage.

| Name of the Fruit | Number of students who like it |
| --- | --- |
| Mango | 65 |
| Grapes | 30 |
| Banana | 5 |
| Other | 20 |



```python
import matplotlib.pyplot as plt

fruits = ['Mango', 'Grapes', 'Banana', 'Other']
counts = [65, 30, 5, 20]

fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(counts, labels=fruits, autopct='%1.1f%%', startangle=90,
       colors=['#ffbe0b', '#8338ec', '#fb5607', '#3a86ff'])
ax.set_title('Fruits Preferred by Students (Pie Chart)')
ax.axis('equal')
plt.show()
```

### 3.

Draw 3D wireframe plot of the function f(x,y) = – x3 – y 3 in Brown color for, x, y  [–15, 15]


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

x = np.linspace(-15, 15, 100)
y = np.linspace(-15, 15, 100)
X, Y = np.meshgrid(x, y)
Z = -(X**3 + Y**3)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='brown', rstride=4, cstride=4)
ax.set_title(r'Wireframe of $f(x, y) = -(x^3 + y^3)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```

### 4.

Using sympy declare quadrilateral ABCD with vertices A (6,2) , B (4,8), C(12,9) and

D(5,3).

Find the area and perimeter of ABCD.


```python
from sympy import Point, Polygon

A = Point(6, 2)
B = Point(4, 8)
C = Point(12, 9)
D = Point(5, 3)

quadrilateral = Polygon(A, B, C, D)
print(f'Area of quadrilateral ABCD: {quadrilateral.area}')
print(f'Perimeter of quadrilateral ABCD: {quadrilateral.perimeter}')
```

### 5.

Using sympy declare  PQR where P (1 , 1), Q(5, 1), R( 5, 5) ,for this triangle find its area,
circumcircle, Is it right angled triangle ? Scale  PQR in x and y directions by factors 2 and 4
respectively .


```python
from sympy import Point, Triangle

P = Point(1, 1)
Q = Point(5, 1)
R = Point(5, 5)

triangle = Triangle(P, Q, R)
print(f'Area of triangle PQR: {triangle.area}')
print(f'Circumcircle of triangle PQR: {triangle.circumcircle}')
print(f'Is triangle PQR right angled? {triangle.is_right()}')

scaled_points = [Point(pt.x * 2, pt.y * 4) for pt in (P, Q, R)]
for label, pt in zip(["P'", "Q'", "R'"], scaled_points):
    print(f'Scaled point {label}: {pt}')
```

### Q2.1 (Attempt Any One)

1. Write a python program to plot the polygon with vertices at points (1, 2) , (4 2), (4 4) and
(2,5). Plot shearing of this polygon shearing in x and y direction by 3 and -2 units
respectively. Also plot x axis and y axis in black colours.

```python
import numpy as np
import matplotlib.pyplot as plt

polygon = np.array([[1, 2], [4, 2], [4, 4], [2, 5]], dtype=float)

def apply_transform(matrix, pts):
    return np.array([matrix @ p for p in pts])

shear_x = np.array([[1, 3], [0, 1]], dtype=float)
shear_y = np.array([[1, 0], [-2, 1]], dtype=float)

sheared_x = apply_transform(shear_x, polygon)
sheared_y = apply_transform(shear_y, polygon)

def close_loop(pts):
    return np.vstack([pts, pts[0]])

plt.figure(figsize=(6, 6))
plt.plot(*close_loop(polygon).T, color='purple', marker='o', label='Original Polygon')
plt.plot(*close_loop(sheared_x).T, color='orange', marker='s', label='Shear in x (k=3)')
plt.plot(*close_loop(sheared_y).T, color='green', marker='^', label='Shear in y (k=-2)')

plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.title('Polygon Shearing Transformations')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle=':')
plt.show()
```

2. Write a python program to generate and plot 80 points on parabola y2 = 16 x using iterative
formulae for 0  x  80,
.

```python
import matplotlib.pyplot as plt

num_points = 80
max_x = 80
step = max_x / (num_points - 1)

xs = []
ys = []
x_val = 0.0
for _ in range(num_points):
    y_val = (16 * x_val) ** 0.5
    xs.append(x_val)
    ys.append(y_val)
    x_val += step

plt.figure(figsize=(7, 4))
plt.scatter(xs, ys, color='crimson', label=r'$y^2 = 16x$')
plt.title(r'Points on Parabola $y^2 = 16x$ for $0 \leq x \leq 80$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```
