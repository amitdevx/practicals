Represent the following information using a Pie chart. Use explode function for separate the sectors.

### 3.

Draw Surface plot for the function f(x, y) = 𝑒 (x +y ) when –2𝜋 x, y 2𝜋, with red
colour.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

grid = np.linspace(-2 * np.pi, 2 * np.pi, 150)
X, Y = np.meshgrid(grid, grid)
Z = np.exp(X**2 + Y**2)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, color="red", edgecolor="none", alpha=0.8)
ax.set_title(r"Surface Plot of $f(x, y) = e^{x^2 + y^2}$")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
plt.show()
```

### 4.

Using sympy declare line segment passing through points P(4, –5) and Q(–7,6). Apply each of
the given transformation on segment PQ. a) reflection through the line y = –x,
b) Translation in x and y direction by factor 2 units and -3 units respectively.


```python
from sympy import Point

P = Point(4, -5)
Q = Point(-7, 6)

def reflect_y_eq_neg_x(point):
    return Point(-point.y, -point.x)

reflected_P = reflect_y_eq_neg_x(P)
reflected_Q = reflect_y_eq_neg_x(Q)

translated_P = Point(P.x + 2, P.y - 3)
translated_Q = Point(Q.x + 2, Q.y - 3)

print(f"Original segment PQ: P{P}, Q{Q}")
print(f"Reflection about y = -x: P{reflected_P}, Q{reflected_Q}")
print(f"Translation (+2, -3): P{translated_P}, Q{translated_Q}")
```

### 5.

Using sympy declare polygon with vertices A (4 , 3), B(–4, 3), C(– 4, –3) and D(4, –3) ,for this
polygon find its area, perimeter, centroid and all angles.


```python
from sympy import Point, Polygon, pi

polygon = Polygon(Point(4, 3), Point(-4, 3), Point(-4, -3), Point(4, -3))

area = polygon.area
perimeter = polygon.perimeter
centroid = polygon.centroid
angles = {vertex: float(angle.evalf() * 180 / pi) for vertex, angle in polygon.angles.items()}

print(f"Polygon vertices: {polygon.vertices}")
print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
print(f"Centroid: {centroid}")
for vertex, angle in angles.items():
    print(f"Angle at {vertex}: {angle:.2f} degrees")
```

### Q2.1 (Attempt Any One)

1. Write a python program to plot the Pyramid OABC with the vertices O(0, 0, 0), A(1, 1, 0),
B(0, 1, 0) and C(0, 0, 1). Find and plot its
1) rotation about x-axis through angle 900 ,
2) reflection through YZ-plane .

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401
from itertools import combinations

vertices = {
    "O": np.array([0, 0, 0]),
    "A": np.array([1, 1, 0]),
    "B": np.array([0, 1, 0]),
    "C": np.array([0, 0, 1]),
}

def rotation_matrix_x(angle_rad):
    return np.array(
        [
            [1, 0, 0],
            [0, np.cos(angle_rad), -np.sin(angle_rad)],
            [0, np.sin(angle_rad), np.cos(angle_rad)],
        ]
    )

rot_x_90 = rotation_matrix_x(np.deg2rad(90))
reflection_yz = np.diag([-1, 1, 1])

rotated_vertices = {k: rot_x_90 @ v for k, v in vertices.items()}
reflected_vertices = {k: reflection_yz @ v for k, v in vertices.items()}

edges = list(combinations(vertices.keys(), 2))

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection="3d")

def plot_shape(vertex_dict, color, label):
    for start, end in edges:
        xs = [vertex_dict[start][0], vertex_dict[end][0]]
        ys = [vertex_dict[start][1], vertex_dict[end][1]]
        zs = [vertex_dict[start][2], vertex_dict[end][2]]
        ax.plot(xs, ys, zs, color=color, linewidth=2)
    centroid = np.mean(list(vertex_dict.values()), axis=0)
    ax.scatter(*centroid, color=color, label=label, s=50)

plot_shape(vertices, "blue", "Original OABC")
plot_shape(rotated_vertices, "green", "Rotated about X-axis (90°)")
plot_shape(reflected_vertices, "red", "Reflection in YZ-plane")

ax.set_title("Transformations of Pyramid OABC")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()
plt.show()
```

2. Write a Python program to generate and plot 50 points on parabola with equation y2 = 20x,
where 0 ≤ 𝑥 ≤ 40 using iterative formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

num_points = 50
t_min = -np.sqrt(20 * 40)
t_max = np.sqrt(20 * 40)
t_values = np.linspace(t_min, t_max, num_points)

points = np.array([(t**2 / 20, t) for t in t_values if 0 <= t**2 / 20 <= 40])

plt.figure(figsize=(7, 5))
plt.scatter(points[:, 0], points[:, 1], color="darkcyan", label=r"$y^2 = 20x$")
plt.title("Iterative Points on Parabola")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0, 40)
plt.grid(True, linestyle=":")
plt.legend()
plt.show()
```
