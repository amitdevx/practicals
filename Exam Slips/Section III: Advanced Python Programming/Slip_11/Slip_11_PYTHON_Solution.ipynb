**Q1.** List of 10 integers. Display square using lambda and map.


```python
l = list(range(1, 11))
print(f"Original list: {l}")

sq = list(map(lambda x: x**2, l))
print(f"Squares: {sq}")
```

**Q2 (OR Option).** String operations: split, find non-vowel words, sort, count.


```python
s = input("Enter a string: ")
if not s: s = "Python is a powerful programming language"

words = s.split()
vowels = "aeiouAEIOU"

no_vowel_words = [w for w in words if not any(char in vowels for char in w)]

no_vowel_words_sorted = sorted(no_vowel_words, key=str.lower)

print(f"Original String: {s}")
print(f"Words without vowels (sorted): {no_vowel_words_sorted}")
print(f"Total count: {len(no_vowel_words_sorted)}")
```
