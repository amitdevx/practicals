Q1. Write a program to accept a string, extract words starting with capital letter, remove special characters.
[10 Marks]


```python
import re
text = input('Enter a string: ') or 'Hello World! From Python Programming.'
words_with_caps = [word for word in text.split() if word[0].isupper()]
clean_text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
print('Words starting with capital:', words_with_caps)
print('Cleaned text:', clean_text)
```

Q2 (OR). Write a program to copy file contents from one file to another.
[20 Marks]


```python
import os
source_file = input('Enter source filename: ') or 'test.txt'
if not os.path.exists(source_file):
    print(f'File {source_file} does not exist')
else:
    with open(source_file, 'r') as src:
        content = src.read()
    dest_file = 'copy_' + source_file
    with open(dest_file, 'w') as dst:
        dst.write(content)
    lines = len(content.splitlines())
    words = len(content.split())
    chars = len(content)
    print(f'File copied to {dest_file}')
    print(f'Lines: {lines}, Words: {words}, Characters: {chars}')
```
