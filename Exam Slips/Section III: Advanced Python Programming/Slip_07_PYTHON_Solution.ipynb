Q1. Write a program to accept a string, extract words starting with capital letter, remove special characters.
[10 Marks]


```python
import re

try:
    text = input('Enter a string: ')
except Exception:
    text = 'Hello World! From Python Programming.'
    print(f"Input: {text}")

if not text:
    text = 'Hello World! From Python Programming.'

word_count = len(text.split())
char_count = len(text)
vowel_count = len(re.findall(r'[aeiouAEIOU]', text))

print(f'Word count: {word_count}')
print(f'Character count: {char_count}')
print(f'Vowel count: {vowel_count}')
```

Q2 (OR). Write a program to copy file contents from one file to another.
[20 Marks]


```python
import os
import shutil

try:
    source_file = input('Enter source filename: ')
except Exception:
    source_file = 'test.txt'
    print(f"Input: {source_file}")

if not source_file:
    source_file = 'test.txt'

dest_file = source_file.replace('.', '_backup.')

# Create a test file if it doesn't exist
if not os.path.exists(source_file):
    with open(source_file, 'w') as f:
        f.write('This is a test file for backup demonstration.')
    print(f'Created test file: {source_file}')

shutil.copy2(source_file, dest_file)
print(f'Backed up {source_file} to {dest_file}')
print(f'Source size: {os.path.getsize(source_file)} bytes')
print(f'Backup size: {os.path.getsize(dest_file)} bytes')
```
