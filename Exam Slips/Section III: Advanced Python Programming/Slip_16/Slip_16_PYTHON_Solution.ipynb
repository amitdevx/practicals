Q2. Write a python program to create a list, string and tuple in python. Apply random module
    to display unique multiple random elements from above created sequence type.

```python
import random

random.seed(77)
numbers_list = list(range(1, 16))
phrase = "AdvancedPython"
cities = ("Pune", "Mumbai", "Satara", "Nashik", "Kolhapur", "Nagpur")

print("Unique random elements from list:", random.sample(numbers_list, 5))
print("Unique random elements from string:", random.sample(sorted(set(phrase)), 4))
print("Unique random elements from tuple:", random.sample(cities, 3))
```

Q2. Write a python program using user define function to accept a number and perform the following
    task:
     a) Check number is Armstrong or not.
     b) Calculate and display sum of digits till single digit [e.g. - 558 = 18 = 9]

```python
def is_armstrong(number: int) -> bool:
    digits = [int(d) for d in str(number)]
    power = len(digits)
    return number == sum(d ** power for d in digits)

def digit_root(number: int) -> int:
    while number >= 10:
        number = sum(int(d) for d in str(number))
    return number

value = 407
print(f"Is {value} an Armstrong number?", is_armstrong(value))
print(f"Digit root of {value}:", digit_root(value))
```
