## Q2. Create a package vehicle with a sub package types, sub package type contains modules cars and
bikes. Each module should have a function available_models() returning a list of models. Write a
python program to display all available models of cars and bikes.
[20 Marks]

```python
df = pd.DataFrame(data)

base_dir = "vehicle"

types_dir = os.path.join(base_dir, "types")

os.makedirs(types_dir, exist_ok=True)

open(os.path.join(base_dir, "__init__.py"), 'w').close()

open(os.path.join(types_dir, "__init__.py"), 'w').close()

with open(os.path.join(types_dir, "cars.py"), "w") as f:
    f.write("def available_models():\n    return ['Sedan', 'SUV', 'Hatchback']")

with open(os.path.join(types_dir, "bikes.py"), "w") as f:
    f.write("def available_models():\n    return ['Sports', 'Cruiser', 'Street']")

sys.path.append(os.getcwd())
```
