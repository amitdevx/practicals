## Q1. Write a python program to accept a string and substring from user. Check if a particular substring is
present at the beginning of the given string, give appropriate message. (Use RE module)
[10 Marks]

```python
import re

text = "Savitribai Phule Pune University inspires many learners."
substring = "Savitribai"
pattern = re.compile(rf"^{re.escape(substring)}", re.IGNORECASE)
if pattern.search(text):
    print(f"'{substring}' is at the beginning of the string.")
else:
    print(f"'{substring}' is NOT at the beginning of the string.")
```

## Q2. Write a Python program to accept a file name and a byte position as input. If file exist open a file in
read mode and move the file pointer to the given byte position, prints the content of the file from
that position till end of file otherwise give appropriate message.
[20 Marks]

```python
from pathlib import Path

slip_dir = Path(__file__).resolve().parent if '__file__' in globals() else Path.cwd()
file_path = slip_dir / "byte_demo.txt"
file_path.write_text("Python makes file handling approachable.\nLet us explore byte offsets!", encoding="utf-8")
byte_position = 18

if file_path.exists():
    with file_path.open("r", encoding="utf-8") as handle:
        handle.seek(byte_position)
        content = handle.read()
        print(f"Reading {file_path.name} from byte {byte_position} onwards:\n{content}")
else:
    print(f"File {file_path.name} does not exist.")
```

## Q2. Write a python program to create a module containing factorial function, power dictionary and list
of vowels. Save this as variables.py. Import variables module to print the factorial of n, print the
power of n, and second alphabet from the vowels list. (Accept n from user)

```python
from pathlib import Path
import importlib.util
import sys

slip_dir = Path(__file__).resolve().parent if '__file__' in globals() else Path.cwd()
module_path = slip_dir / "variables.py"
module_code = """from math import factorial as math_factorial

def factorial(n):
    return math_factorial(n)

power = {i: i ** 2 for i in range(1, 21)}
vowels = ['a', 'e', 'i', 'o', 'u']
"""
module_path.write_text(module_code, encoding="utf-8")

spec = importlib.util.spec_from_file_location("variables", module_path)
variables = importlib.util.module_from_spec(spec)
sys.modules["variables"] = variables
spec.loader.exec_module(variables)

n = 5
print(f"Factorial of {n} is {variables.factorial(n)}")
print(f"Square of {n} from power dictionary is {variables.power.get(n)}")
print(f"Second alphabet from vowels list: {variables.vowels[1]}")
```
