Q1. Write a Python program to shuffle a deck of card. Display the cards obtained by a user in 5
    attempts.
                                                                                              [10 Marks]


```python
import random

# Q1: Shuffle a deck of cards
suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
ranks = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King', 'Ace']
deck = [f"{rank} of {suit}" for suit in suits for rank in ranks]
random.shuffle(deck)

print("Your 5 cards:")
for i in range(5):
    print(deck[i])
```

Q2. Create a package finance_tools with modules interest.py and currency.py
     a) interest.py should contain function cal_interest(amount, year) that accepts amount, year . Calculates
        and returns simple interest using a fixed rate (for example, 5%).
     b) currency.py should contain function converet(amount) that accepts amount in INR and converts it into
        USD. (consider conversion rate 0.012)
    Write a python program to import modules from finance_tools package and demonstrate both
    functions.
                                                                                              [20 Marks]

```python
# Q2: Create finance_tools package
base_dir = "finance_tools"
if not os.path.exists(base_dir):
    os.makedirs(base_dir)
    open(os.path.join(base_dir, "__init__.py"), 'a').close()

    with open(os.path.join(base_dir, "interest.py"), "w") as f:
        f.write("def cal_interest(amount, year, rate=5):\n    return (amount * year * rate) / 100")

    with open(os.path.join(base_dir, "currency.py"), "w") as f:
        f.write("def convert(amount):\n    # 1 INR = 0.012 USD\n    return amount * 0.012")

    print(f"Package 'finance_tools' created at {os.path.abspath(base_dir)}")
else:
    print("Package 'finance_tools' already exists.")

sys.path.append(os.getcwd())

try:
    from finance_tools import interest, currency

    amt = 10000
    years = 2
    si = interest.cal_interest(amt, years)
    usd = currency.convert(amt)

    print(f"Principal: {amt}, Years: {years}")
    print(f"Simple Interest (5%): {si}")
    print(f"Amount in USD (rate 0.012): {usd}")

except ImportError as e:
    print(f"Error importing package: {e}")
```

Q2. Write a python program to check weather given url is valid or not?
    (for Ex: https://www.yahoo.com)

```python
import re

# Q2 OR: URL validation
def is_valid_url(url):
    regex = r"^(http|https)://[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}(/.*)?$"
    if re.match(regex, url):
        return True
    return False

urls = [
    "https://www.yahoo.com",
    "http://google.com",
    "ftp://example.com",
    "www.yahoo.com",
    "https://site",
    "invalid-url"
]

try:
    user_url = input("Enter a URL to check: ")
    if user_url:
        urls.insert(0, user_url)
except EOFError:
    pass

for url in urls:
    if is_valid_url(url):
        print(f"'{url}' is Valid")
    else:
        print(f"'{url}' is Invalid")

# DataFrame operations
data = {
    'Company': ['Apsara', 'Natraj', 'Cello', 'Parkar', 'Apsara'],
    'Count': [15, 20, 25, 35, 20],
    'Price': [250, 200, 600, 900, 300]
}
labels = ['Pencil', 'Pencil', 'Pen', 'Pen', 'Eraser']
df = pd.DataFrame(data, index=labels)

print("\nOriginal DataFrame:")
print(df)

print("\n-- a) Rows with label 'Pencil' --")
print(df.loc['Pencil'])

print("\n-- b) Update Eraser Count --")
df.loc['Eraser', 'Count'] = 25
print(df.loc['Eraser'])

print("\n-- c) Columns Company and Price --")
print(df[['Company', 'Price']])

print("\n-- d) Rows 'Pencil' and 'Pen' --")
print(df.loc[['Pencil', 'Pen']])

print("\n-- e) Add 'Colour' Column --")
df['Colour'] = ['Black', 'Red', 'Blue', 'Green', 'White']
print(df)
```
