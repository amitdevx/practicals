Q2. Write a Python program that calculates the pass or fail status of multiple students.
        a) Define a function check_pass_fail(marks) to return "Pass" if marks ≥ 40, otherwise "Fail".
        b) Accept number of students and their marks.
        c) Display each student’s result.
        d) Print the total count of students who passed and failed.
                                                                                           [20 Marks]

```python
# Q2: Student marks - pass/fail
try:
    n_str = input("Enter number of students: ")
    n = int(n_str)
except Exception:
    n = 3
    print(f"Input: n={n}")

student_marks = {}
for i in range(n):
    try:
        marks_str = input(f"Enter marks for Student {i+1}: ")
        marks = float(marks_str)
    except Exception:
        marks = 45.0 + i * 15  # Demo: 45, 60, 75
        print(f"Input: marks={marks}")
    student_marks[f"Student {i+1}"] = marks

pass_count = sum(1 for m in student_marks.values() if m >= 40)
fail_count = len(student_marks) - pass_count

print(f"\nResults:")
for name, marks in student_marks.items():
    status = "Pass" if marks >= 40 else "Fail"
    print(f"  {name}: {marks} - {status}")
print(f"\nTotal Pass: {pass_count}, Total Fail: {fail_count}")
```

Q2. Write a Python program that:
       a) Retrieves current local time and UTC time.
       b) Prints the current time using ctime() and gmtime().
       c) Accepts a date-time string input in the format DD-MM-YYYY HH:MM:SS.
       d) Parses the string using strptime() and formats it back using strftime() and display it in two
           custom formats.

```python
from datetime import datetime

# Q2 OR: Show date-time and day info
try:
    dt_str = input("Enter date-time (DD-MM-YYYY HH:MM:SS): ")
    dt = datetime.strptime(dt_str, "%d-%m-%Y %H:%M:%S")
except Exception:
    dt = datetime.now()
    print(f"Using current date-time: {dt.strftime('%d-%m-%Y %H:%M:%S')}")

print(f"Day of Week: {dt.strftime('%A')}")
print(f"Day of Month: {dt.day}")
print(f"Day of Year: {dt.timetuple().tm_yday}")
```
