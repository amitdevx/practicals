## Q1. Write a Python program that accepts value of m and n. Divides m by n and handles a
ZeroDivisionError if the user tries to divide by zero.
[10 Marks]


```python
import os

try:
    m = float(input("Enter m (numerator): "))
    n = float(input("Enter n (denominator): "))
except Exception:
    m, n = 10.0, 3.0
    print(f"Input: m={m}, n={n}")

try:
    result = m / n
    print(f"Result: {result}")
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
```

## Q2. Write a python program to accept a file name from user. If file exist display contents of file and
number of characters, words and lines in a file, otherwise display appropriate message.

```python
try:
    filename = input("Enter filename to check: ")
except Exception:
    filename = "variables.py"
    print(f"Input: {filename}")

if os.path.exists(filename):
    try:
        with open(filename, 'r') as f:
            content = f.read()

        num_chars = len(content)
        num_words = len(content.split())
        num_lines = content.count('\n') + (1 if content else 0)

        print(f"\n--- Content of {filename} ---")
        print(content)
        print(f"\nNumber of characters: {num_chars}")
        print(f"Number of words: {num_words}")
        print(f"Number of lines: {num_lines}")

    except Exception as e:
        print(f"Error reading file: {e}")
else:
    print("File does not exist.")
```
