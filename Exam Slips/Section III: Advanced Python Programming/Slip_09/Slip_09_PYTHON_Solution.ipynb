```python
def even_generator(n):
    for i in range(n + 1):
        if i % 2 == 0:
            yield i

student1_data = {
    'ID': [0, 1, 2, 3, 4],
    'Name': ['S2', 'S4', 'S5', 'S6', 'S7'],  # Based on table headers: ID, City, Marks?

}

df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'City': ['Pune', 'Nashik', 'Mumbai', 'Delhi', 'Nanded'],
    'Marks': [97, 78, 86, 66, 54]
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3, 6, 7],
    'Name': ['Amit', 'Bob', 'Charlie', 'Dave', 'Eve'],
    'Age': [20, 21, 22, 23, 24]
})

df3 = pd.merge(df1, df2, on='ID', how='inner')

def is_armstrong(num):
    s = str(num)
    n = len(s)
    total = sum(int(digit) ** n for digit in s)
    return total == num

def sum_digits_single(num):
    while num > 9:
        num = sum(int(d) for d in str(num))
    return num
```
