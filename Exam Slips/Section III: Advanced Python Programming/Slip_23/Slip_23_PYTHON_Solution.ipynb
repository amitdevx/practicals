**Q1.** Extract number from string.


```python
import re

s = input("Enter a string: ")
if not s: s = "Order ID is 12345 and Price is 500"

nums = re.findall(r'\d+', s)
print(f"String: {s}")
print(f"Extracted numbers: {nums}")
```

**Q2.** Calculate pass/fail status.


```python
def check_pass_fail(marks):
    return "Pass" if marks >= 40 else "Fail"

try:
    n_input = input("Enter number of students: ")
    n = int(n_input) if n_input else 3
    
    results = []
    pass_count = 0
    fail_count = 0
    
    dummy_marks = [35, 80, 45]

    for i in range(n):
        if n_input:
             m = float(input(f"Enter marks for student {i+1}: "))
        else:
             m = dummy_marks[i]
             print(f"Marks for student {i+1}: {m}")
             
        status = check_pass_fail(m)
        if status == "Pass":
            pass_count += 1
        else:
            fail_count += 1
        results.append(status)
        print(f"Student {i+1} Result: {status}")

    print("-" * 20)
    print(f"Total Pass: {pass_count}")
    print(f"Total Fail: {fail_count}")

except ValueError:
    print("Invalid input")
```
