Q1. Create a data.csv file containing employee name, designation, salary of an employee. Write a
python program to create a DataFrame using data.csv and display the contents of DataFrame with
statistical information of dataframe.
[10 Marks]

```python
from pathlib import Path
import pandas as pd

slip_dir = Path(__file__).resolve().parent if '__file__' in globals() else Path.cwd()
csv_path = slip_dir / "data.csv"
records = [
    {"name": "Asha Kulkarni", "designation": "Data Analyst", "salary": 52000},
    {"name": "Nikhil Deshmukh", "designation": "Software Engineer", "salary": 68000},
    {"name": "Priya Patil", "designation": "Project Manager", "salary": 91000},
    {"name": "Sohan Kamat", "designation": "QA Lead", "salary": 58000},
]
df = pd.DataFrame(records)
df.to_csv(csv_path, index=False)
print(f"CSV file created at: {csv_path}")
print("\nDataFrame contents:\n", df, sep="")
print("\nStatistical summary:\n", df.describe(include='all'), sep="")
```

Q2. Write a Python program that:
a) Retrieves current local time and UTC time.
b) Prints the current time using ctime() and gmtime().
c) Accepts a date-time string input in the format DD-MM-YYYY HH:MM:SS.
d) Parses the string using strptime() and formats it back using strftime() and display it in two
custom formats.
[20 Marks]

### OR

```python
import time
from datetime import datetime

current_epoch = time.time()
local_time = time.localtime(current_epoch)
utc_time = time.gmtime(current_epoch)
print("Local time:", time.strftime("%Y-%m-%d %H:%M:%S", local_time))
print("UTC time:", time.strftime("%Y-%m-%d %H:%M:%S", utc_time))
print("ctime():", time.ctime(current_epoch))
print("gmtime() as string:", time.asctime(utc_time))

input_string = "24-01-2025 11:30:45"
parsed_datetime = datetime.strptime(input_string, "%d-%m-%Y %H:%M:%S")
print("\nParsed datetime:", parsed_datetime)
print("Format 1:", parsed_datetime.strftime("%B %d, %Y at %I:%M %p"))
print("Format 2:", parsed_datetime.strftime("Week %U | %A %H:%M"))
```

Q2. Write a python program to create a list, string and tuple in python. Apply random module to display
unique multiple random elements from above created sequence type.

```python
import random

random.seed(27)
numbers_list = list(range(10, 21))
sample_string = "LearningPythonIsFun"
city_tuple = ("Pune", "Mumbai", "Nashik", "Nagpur", "Kolhapur", "Satara")

print("Numbers:", numbers_list)
print("Random unique sample:", random.sample(numbers_list, 4))

unique_chars = sorted(set(sample_string))
print("String:", sample_string)
print("Random unique characters:", random.sample(unique_chars, 5))

print("Tuple:", city_tuple)
print("Random unique cities:", random.sample(city_tuple, 3))
```
