c) List only the columns Company and Price.
d) List only rows with labels ‘Pencil’ and ‘Pen’
e) Rename Column Count as Quantity and display DataFrame.
[20 Marks]

```python
# Q2 - DataFrame operations
data = {
    'Company': ['Apsara', 'Natraj', 'Cello', 'Parkar', 'Apsara'],
    'Count': [15, 20, 25, 35, 20],
    'Price': [250, 200, 600, 900, 300]
}

index_labels = ['Pencil', 'Pencil', 'Pen', 'Pen', 'Eraser']

df = pd.DataFrame(data, index=index_labels)

print("Original DataFrame:")
print(df)

print("\na) Find all rows with the label 'Pencil':")
print(df.loc['Pencil'])

print("\nb) Change the Eraser count as 25 instead of 20:")
df.loc['Eraser', 'Count'] = 25
print(df)

print("\nc) List only the columns Company and Price:")
print(df[['Company', 'Price']])

print("\nd) List only rows with labels 'Pencil' and 'Pen':")
print(df.loc[['Pencil', 'Pen']])

print("\ne) Rename Column Count as Quantity:")
df.rename(columns={'Count': 'Quantity'}, inplace=True)
print(df)
```

## Q2. Write a python program for following
a) Create a file which stores the first name, middle name and last name of your 5 friends
b) Display contents of file
c) Append other 5 friend’s names
d) Display modified file contents

```python
# Q2 (OR) - File operations with friends' names
filename = "friends.txt"

# a) Create file with first name, middle name and last name of 5 friends
print("a) Creating file with 5 friends' names...")
friends_initial = [
    "Alice Marie Johnson",
    "Bob James Smith",
    "Charlie Ray Davis",
    "David Paul Wilson",
    "Eve Lynn Brown"
]
with open(filename, 'w') as f:
    for friend in friends_initial:
        f.write(friend + "\n")

# b) Display contents of file
print("b) Display contents of file:")
with open(filename, 'r') as f:
    print(f.read())

# c) Append other 5 friends' names
print("c) Appending other 5 friends' names...")
friends_append = [
    "Frank Michael Lee",
    "Grace Ann Taylor",
    "Hank Joseph Miller",
    "Ivy Rose Anderson",
    "Jack Daniel Thomas"
]
with open(filename, 'a') as f:
    for friend in friends_append:
        f.write(friend + "\n")

# d) Display modified file contents
print("d) Display modified file contents:")
with open(filename, 'r') as f:
    print(f.read())

if os.path.exists(filename):
    os.remove(filename)
```
