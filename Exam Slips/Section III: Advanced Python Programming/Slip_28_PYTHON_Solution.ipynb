Q1. Write a Python program using regular expressions to validate whether a given string matches the
    pattern of three-letter words such as cat, cot, cut, mat, met, mud.
                                                                                            [10 Marks]


```python
import re
import time
from datetime import datetime

# Q1: Regex - three-letter words
try:
    text = input("Enter a string: ")
except EOFError:
    text = "The cat cut the cot and met the mud"
    print(f"Input: {text}")

words = text.split()
pattern = r"\b[a-zA-Z]{3}\b"
matching_words = [word for word in words if re.fullmatch(pattern, word)]
print(f"Three-letter words found: {matching_words}")
```

Q2. Build a package education with a sub package courses, containing modules science.py and

    arts.py. Each module should have a function get_subjects(). Write a python program to display all
    subjects.
                                                                                            [20 Marks]

```python
import os
import sys

# Q2: Build education package
def create_education_package():
    base_dir = "education"
    sub_dir = os.path.join(base_dir, "courses")
    if not os.path.exists(sub_dir):
        os.makedirs(sub_dir)
        open(os.path.join(base_dir, "__init__.py"), 'a').close()
        open(os.path.join(sub_dir, "__init__.py"), 'a').close()

        with open(os.path.join(sub_dir, "science.py"), "w") as f:
            f.write("def get_subjects():\n    return ['Physics', 'Chemistry', 'Biology']")

        with open(os.path.join(sub_dir, "arts.py"), "w") as f:
            f.write("def get_subjects():\n    return ['History', 'Geography', 'Political Science']")

        print("Created package 'education'")

create_education_package()
sys.path.append(os.getcwd())

try:
    from education.courses import science, arts
    print("Science Subjects:", science.get_subjects())
    print("Arts Subjects:", arts.get_subjects())
except ImportError as e:
    print(f"Error: {e}")
```

Q2. Write a python program to display all entries in a current directory and identify each as file or
    folder. [Use getcwd() function to retrieve current directory]

```python
current_dir = os.getcwd()
print(f"\nCurrent Directory: {current_dir}")
try:
    entries = os.listdir(current_dir)
    for entry in entries:
        full_path = os.path.join(current_dir, entry)
        if os.path.isfile(full_path):
            print(f"[FILE]   {entry}")
        elif os.path.isdir(full_path):
            print(f"[FOLDER] {entry}")
except Exception as e:
    print(f"Error: {e}")
```
