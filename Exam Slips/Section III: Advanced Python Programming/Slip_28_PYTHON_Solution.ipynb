Q2. Write a Python program that accepts text file name and a string. If file exist, then appends string to
    its content and display the contents of the file otherwise gives appropriate message.
                                                                                            [20 Marks]

```python
# Q2: File append operation
filename = "test_append.txt"
with open(filename, "w") as f:
    f.write("Original Content.\n")

try:
    fname = input("Enter filename (default: test_append.txt): ")
    if not fname: fname = filename

    string_to_append = input("Enter string to append: ")
    if not string_to_append: string_to_append = " Appended Content."

    if os.path.exists(fname):
        with open(fname, 'a') as f:
            f.write(string_to_append)
        print("\nUpdated File Content:")
        with open(fname, 'r') as f:
            print(f.read())
    else:
        print("File does not exist.")
except Exception as e:
    print(f"Error: {e}")

if os.path.exists(filename):
    os.remove(filename)
```

Q2. Write a Python program to validate mobile number based on following criteria
       a) The first digit should contain numbers between 6 to 9.
       b) The rest 9 digit can contain any number between 0 to 9.
       c) The mobile number can have 11 digits also by including 0 at the starting.
       d) The mobile number can be of 12 digits also by including 91 at the starting

```python
# Q2 OR: Mobile number validation
def validate_mobile(number):
    pattern = r"^(0|91)?[6-9]\d{9}$"
    if re.match(pattern, number):
        return True
    return False

try:
    num = input("Enter mobile number: ")
    if not num: num = "9876543210"
except EOFError:
    num = "9876543210"

if validate_mobile(num):
    print("Valid Mobile Number")
else:
    print("Invalid Mobile Number")

# Education package
def create_education_package():
    base_dir = "education"
    sub_dir = os.path.join(base_dir, "courses")
    if not os.path.exists(sub_dir):
        os.makedirs(sub_dir)
        open(os.path.join(base_dir, "__init__.py"), 'a').close()
        open(os.path.join(sub_dir, "__init__.py"), 'a').close()

        with open(os.path.join(sub_dir, "science.py"), "w") as f:
            f.write("def get_subjects():\n    return ['Physics', 'Chemistry', 'Biology']")

        with open(os.path.join(sub_dir, "arts.py"), "w") as f:
            f.write("def get_subjects():\n    return ['History', 'Geography', 'Political Science']")

        print("Created package 'education'")

create_education_package()
sys.path.append(os.getcwd())

try:
    from education.courses import science, arts
    print("Science Subjects:", science.get_subjects())
    print("Arts Subjects:", arts.get_subjects())
except ImportError as e:
    print(f"Error: {e}")

# Directory listing
current_dir = os.getcwd()
print(f"\nCurrent Directory: {current_dir}")
try:
    entries = os.listdir(current_dir)
    for entry in entries:
        full_path = os.path.join(current_dir, entry)
        if os.path.isfile(full_path):
            print(f"[FILE]   {entry}")
        elif os.path.isdir(full_path):
            print(f"[FOLDER] {entry}")
        else:
            print(f"[OTHER]  {entry}")
except Exception as e:
    print(f"Error: {e}")
```
