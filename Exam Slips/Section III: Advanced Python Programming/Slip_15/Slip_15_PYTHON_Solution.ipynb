**Q1.** Time module functions.


```python
import time

now = time.time()
print(f"Current time (seconds since epoch): {now}")

struct_utc = time.gmtime(now)
print(f"Struct Time (UTC): {struct_utc}")

formatted = time.strftime("%Y-%m-%d %H:%M:%S", struct_utc)
print(f"Formatted Time: {formatted}")
```

**Q2 (OR Option).** Check if URL is valid.


```python
import re

url = input("Enter URL: ")
if not url: url = "https://www.google.com"

regex = r"^(http|https)://[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}(?:/[a-zA-Z0-9-._~:/?#[\]@!$&'()*+,;=]*)?$"

if re.match(regex, url):
    print(f"'{url}' is a VALID URL")
else:
    print(f"'{url}' is INVALID URL")
```
