## Q2. Write a Python program that accepts text file name and a string. If file exist, then appends string to
    its content and display the contents of the file otherwise gives appropriate message.
                                                                                            [20 Marks]

```python
import os

# Q2: File append operations
try:
    fname = input("Enter filename (default: test_append.txt): ")
    string_to_append = input("Enter string to append: ")
except Exception:
    fname = "test_append.txt"
    string_to_append = "Hello from Slip 29!"
    print(f"Input: filename={fname}, text={string_to_append}")

if not fname:
    fname = "test_append.txt"
if not string_to_append:
    string_to_append = "Default appended text"

with open(fname, 'a') as f:
    f.write(string_to_append + '\n')
print(f"Appended to {fname}")

with open(fname, 'r') as f:
    print(f"\nFile contents:\n{f.read()}")

# Cleanup
if os.path.exists(fname):
    os.remove(fname)
```

## Q2. Write a Python program to validate mobile number based on following criteria
       a) The first digit should contain numbers between 6 to 9.
       b) The rest 9 digit can contain any number between 0 to 9.
       c) The mobile number can have 11 digits also by including 0 at the starting.
       d) The mobile number can be of 12 digits also by including 91 at the starting

```python
# Q2 OR: Mobile number validation
def validate_mobile(number):
    pattern = r"^(0|91)?[6-9]\d{9}$"
    if re.match(pattern, number):
        return True
    return False

try:
    num = input("Enter mobile number: ")
    if not num: num = "9876543210"
except Exception:
    num = "9876543210"

if validate_mobile(num):
    print("Valid Mobile Number")
else:
    print("Invalid Mobile Number")
```
