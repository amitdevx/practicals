Q1. Write a python program to accepts a value of n and create a generator that yields even numbers up
to n.
[10 Marks]


```python
# Q1 OR: Even number generator
def even_generator(n):
    for i in range(2, 2*n+1, 2):
        yield i

try:
    n = int(input("Enter n for even numbers generator: "))
except Exception:
    n = 10
    print(f"Input: n={n}")

print(f"First {n} even numbers:")
for num in even_generator(n):
    print(num, end=' ')
print()
```

```python
# Q2: DataFrames Student 1, Student 2 - merge
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'City': ['Pune', 'Nashik', 'Mumbai', 'Delhi', 'Nanded'],
    'Marks': [97, 78, 86, 66, 54]
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3, 6, 7],
    'Name': ['Amit', 'Bob', 'Charlie', 'Dave', 'Eve'],
    'Age': [20, 21, 22, 23, 24]
})

df3 = pd.merge(df1, df2, on='ID', how='inner')

print("DataFrame Student 1:")
print(df1)
print("\nDataFrame Student 2:")
print(df2)
print("\nMerged DataFrame (Inner Join on ID):")
print(df3)
```
