## Q1. Write a Python program that accepts value of m and n. Divides m by n and handles a
     ZeroDivisionError if the user tries to divide by zero.
                                                                                                [10 Marks]


```python
# Q1 OR: Division with exception handling
try:
    m_str = input("Enter numerator (m): ")
    n_str = input("Enter denominator (n): ")
    m = float(m_str)
    n = float(n_str)
except Exception:
    m, n = 10.0, 3.0
    print(f"Input: m={m}, n={n}")

try:
    result = m / n
    print(f"{m} / {n} = {result}")
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
```

## Q2. Write a python program for following -
      a) Create DataFrames Student 1, Student 2 and display contents
      b) Merge two DataFrames in third frame .
      c) Display contents of third DataFrame.
                        Student 1                                       Student 2

                   ID      Name          Age                     ID       City         Marks
             0      S1     Anita            18               0     S2     Pune             97
             1      S2     Sakshi           23               1     S4     Nashik           78
             2      S3     Om               20               2     S5     Mumbai           86
             3      S4     Raj              21               3     S6     Delhi            66
             4      S5     Kanika           20               4     S7     Nanded           54

                                                                                                [20 Marks]

```python
# Q2: DataFrames Student 1 and 2
data1 = {
    'ID': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'Name': ['Anita', 'Sakshi', 'Om', 'Raj', 'Kanika'],
    'Age': [18, 23, 20, 21, 20]
}
df1 = pd.DataFrame(data1)

data2 = {
    'ID': ['S2', 'S4', 'S5', 'S6', 'S7'],
    'City': ['Pune', 'Nashik', 'Mumbai', 'Delhi', 'Nanded'],
    'Marks': [97, 78, 86, 66, 54]
}
df2 = pd.DataFrame(data2)

df3 = pd.merge(df1, df2, on='ID', how='outer')

print("DataFrame Student 1:")
print(df1)
print("\nDataFrame Student 2:")
print(df2)
print("\nMerged DataFrame (Student 3):")
print(df3)
```

## Q2. Write a Python program to create a class that acts as an iterator and generates the factorial series up
    to n (1! + 2! + 3! + … + n!). (Accept input from the user)

```python
import os
import importlib.util, sys

# Q2 OR: Factorial series using module
module_code = """def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)
"""

module_path = os.path.join(os.getcwd(), "factorial_module.py")
with open(module_path, "w") as f:
    f.write(module_code)

spec = importlib.util.spec_from_file_location("factorial_module", module_path)
mod = importlib.util.module_from_spec(spec)
sys.modules["factorial_module"] = mod
spec.loader.exec_module(mod)

try:
    n_input = input("Enter n for factorial series: ")
    n = int(n_input)
except Exception:
    n = 7
    print(f"Input: n={n}")

print(f"Factorial series up to {n}:")
for i in range(1, n + 1):
    print(f"{i}! = {mod.factorial(i)}")
```
