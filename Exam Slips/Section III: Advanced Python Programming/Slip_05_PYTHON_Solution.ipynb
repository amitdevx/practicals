## Q1. Write a python program to accept a string, find all words that do not contain vowels.
[10 Marks]


```python
import re

try:
    s = input("Enter a string: ")
except Exception:
    s = "Sky Fly Cry Apple Banana Rhythm"
    print(f"Input: {s}")

three_letter = re.findall(r'\b\w{3}\b', s)
print(f"Three-letter words: {three_letter}")
```

## Q2. Write a python program to create a list of prime numbers between m and n using List
comprehension. (Accept the value of m and n from user, m should be less than n)

```python
m = 10  # starting number
n = 50  # ending number
primes = [x for x in range(m, n + 1) if x > 1 and all(x % i != 0 for i in range(2, int(x**0.5) + 1))]
print(f"Prime numbers between {m} and {n}: {primes}")
```
