Q1. Write a python program to accept a string, find all words that do not contain vowels.
[10 Marks]


```python
import os
import sys

s = input("Enter a string: ")
words = s.split()
no_vowel_words = [word for word in words if not any(char.lower() in 'aeiou' for char in word)]
print("Words without vowels:", no_vowel_words)
```

Q2. Create a package banking with modules account.py and loan.py.
a) account.py should contain function open_account(name) that accepts a user’s name and prints a message
confirming account creation
b) loan.py should contain function apply_loan(amount) that accepts a loan amount and prints a confirmation
message with the amount
Write a python program to import modules from banking package and demonstrate both functions.
[20 Marks]

```python
base_dir = "banking"
os.makedirs(base_dir, exist_ok=True)
open(os.path.join(base_dir, "__init__.py"), 'w').close()

with open(os.path.join(base_dir, "account.py"), "w") as f:
    f.write("def open_account(name):\n    print(f'Account created for {name}')")

with open(os.path.join(base_dir, "loan.py"), "w") as f:
    f.write("def apply_loan(amount):\n    print(f'Loan of {amount} applied successfully')")

sys.path.insert(0, os.getcwd())

try:
    from banking import account, loan
    name = input("Enter account holder name: ")
    account.open_account(name)
    amount = float(input("Enter loan amount: "))
    loan.apply_loan(amount)
except Exception as e:
    print(f"Error: {e}")
```

Q2. Write a python program to create a list of prime numbers between m and n using List
comprehension. (Accept the value of m and n from user, m should be less than n)

```python
m = int(input("Enter starting number: "))
n = int(input("Enter ending number: "))
primes = [x for x in range(m, n + 1) if x > 1 and all(x % i != 0 for i in range(2, int(x**0.5) + 1))]
print("Prime numbers:", primes)
```
