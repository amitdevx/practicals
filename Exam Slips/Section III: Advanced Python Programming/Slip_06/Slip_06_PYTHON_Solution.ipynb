**Q1.** Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical.


```python
try:
    val1 = input("Enter number 1: ")
    val2 = input("Enter number 2: ")
    
    if not (val1.replace('.','',1).isdigit() and val2.replace('.','',1).isdigit()):
         raise TypeError("Inputs must be numerical")
    
    n1 = float(val1)
    n2 = float(val2)
    print(f"Numbers are {n1}, {n2}")

except TypeError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Other Error: {e}")
```

**Q2.** Create a following DataFrame named as “data”...
a) Find all rows with the label “Pencil”...
b) Change the count of Eraser...


```python
import pandas as pd

data = {
    'Company': ['Apsara', 'Natraj', 'Cello', 'Parkar', 'Apsara'],
    'Count': [15, 20, 25, 35, 20],
    'Price': [250, 200, 600, 900, 300]
}
index_labels = ['Pencil', 'Pencil', 'Pen', 'Pen', 'Eraser']
df = pd.DataFrame(data, index=index_labels)
print("Original DataFrame:")
print(df)

print("\n-- a) Rows with label 'Pencil' --")
print(df.loc['Pencil'])

print("\n-- b) Change count of Eraser to 25 --")
df.loc['Eraser', 'Count'] = 25
print(df)

print("\n-- c) Columns Company and Price --")
print(df[['Company', 'Price']])

print("\n-- d) Rows with labels 'Pencil' and 'Pen' --")
print(df.loc[['Pencil', 'Pen']])

print("\n-- e) Delete column Count --")
df.drop(columns=['Count'], inplace=True)
print(df)
```
