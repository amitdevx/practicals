Q1. Write a python program to accept a string, Count total number of vowels, consonants and special
    symbols from it.
                                                                                             [10 Marks]

```python
text = "Savitribai pioneered education!"
vowels = set("aeiouAEIOU")
vowel_count = consonant_count = special_count = 0

for ch in text:
    if ch.isalpha():
        if ch in vowels:
            vowel_count += 1
        else:
            consonant_count += 1
    elif not ch.isspace():
        special_count += 1

print("Text:", text)
print(f"Vowels: {vowel_count}")
print(f"Consonants: {consonant_count}")
print(f"Special symbols: {special_count}")
```

Q2. Write a Python program to accept a file name and a byte position as input. If file exist open a file in
    read mode and move the file pointer to the given byte position, prints the content of the file from
    that position till end of file otherwise give appropriate message.
                                                                                            [20 Marks]

### OR

```python
from pathlib import Path

slip_dir = Path(__file__).resolve().parent if '__file__' in globals() else Path.cwd()
file_path = slip_dir / "position_demo.txt"
file_path.write_text("Python makes byte navigation simple.
Track positions easily!", encoding="utf-8")
byte_position = 12

if file_path.exists():
    with file_path.open("r", encoding="utf-8") as handle:
        handle.seek(byte_position)
        data = handle.read()
        print(f"Content from byte {byte_position}:
{data}")
else:
    print(f"File {file_path.name} not found.")
```

Q2. Create a Pandas DataFrame using the following student data with columns and perform the
    following tasks:
           school   class    name     date_of_birth    age   height   weight    address
           S1       S001    Asha      15/05/2002       12    173      35        Street1
           S2       S002    Fransis   17/05/2002       12    192      32        Street2
           S3       S003    Charlie   16/02/1999       13    186      33        Street3
           S4       S004    David     25/09/1998       14    167      30        Street4
           S5       S005    Shyam     11/05/2002       12    151      31        Street5
           S6       S006    Eashel    15/09/1997       16    159      32        Street6
     a) Split the DataFrame into groups based on the school column using the GroupBy operation.
     b) Check and display the type of the GroupBy object created.
     c) Calculate the mean, minimum, and maximum values of the age column for each school group.
     d) Count the number of students whose weight is greater than 32.

```python
import pandas as pd

data = [
    {"school": "S1", "class": "S001", "name": "Asha", "date_of_birth": "15/05/2002", "age": 12, "height": 173, "weight": 35, "address": "Street1"},
    {"school": "S2", "class": "S002", "name": "Fransis", "date_of_birth": "17/05/2002", "age": 12, "height": 192, "weight": 32, "address": "Street2"},
    {"school": "S3", "class": "S003", "name": "Charlie", "date_of_birth": "16/02/1999", "age": 13, "height": 186, "weight": 33, "address": "Street3"},
    {"school": "S4", "class": "S004", "name": "David", "date_of_birth": "25/09/1998", "age": 14, "height": 167, "weight": 30, "address": "Street4"},
    {"school": "S5", "class": "S005", "name": "Shyam", "date_of_birth": "11/05/2002", "age": 12, "height": 151, "weight": 31, "address": "Street5"},
    {"school": "S6", "class": "S006", "name": "Eashel", "date_of_birth": "15/09/1997", "age": 16, "height": 159, "weight": 32, "address": "Street6"},
]

df = pd.DataFrame(data)
grouped = df.groupby('school')
print("Type of GroupBy object:", type(grouped))
print("
Mean/Min/Max age per school:
", grouped['age'].agg(['mean', 'min', 'max']), sep="")
heavier_students = (df['weight'] > 32).sum()
print(f"
Students with weight > 32: {heavier_students}")
```
