**Q1.** Area of circle using function.


```python
def area_circle(r):
    return 3.14 * r * r

try:
    r_input = input("Enter radius: ")
    if not r_input: r_input = "5"
    r = float(r_input)
    print(f"Area of circle with radius {r} is: {area_circle(r)}")
except ValueError:
    print("Invalid input")
```

**Q2.** Create a list, string and tuple. Apply random module to display unique multiple random elements.


```python
import random

my_list = [10, 20, 30, 40, 50, 60, 70]
my_str = "AdvancedPythonProgramming"
my_tuple = (1, 2, 3, 4, 5, 6, 7, 8)

print(f"List: {my_list}")
print(f"String: {my_str}")
print(f"Tuple: {my_tuple}")

k = 3
print(f"\n{k} random unique elements from List: {random.sample(my_list, k)}")
print(f"{k} random unique elements from String: {random.sample(my_str, k)}")
print(f"{k} random unique elements from Tuple: {random.sample(my_tuple, k)}")
```
