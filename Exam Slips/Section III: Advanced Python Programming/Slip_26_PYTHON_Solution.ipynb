Q2. Create a package vehicle with a sub package types, sub package type contains modules cars and
    bikes. Each module should have a function available_models() returning a list of models. Write a
    python program to display all available models of cars and bikes.
                                                                                        [20 Marks]

```python
# Q2: Create vehicle package with sub-package types
def create_vehicle_package():
    base_dir = "vehicle"
    if not os.path.exists(base_dir):
        os.makedirs(base_dir)
        open(os.path.join(base_dir, "__init__.py"), 'a').close()

        sub_dir = os.path.join(base_dir, "types")
        os.makedirs(sub_dir)
        open(os.path.join(sub_dir, "__init__.py"), 'a').close()

        with open(os.path.join(sub_dir, "cars.py"), "w") as f:
            f.write("def available_models():\n    return ['Sedan', 'SUV', 'Hatchback']")

        with open(os.path.join(sub_dir, "bikes.py"), "w") as f:
            f.write("def available_models():\n    return ['Sports', 'Cruiser', 'Dirt Bike']")

        print(f"Package 'vehicle' created at {os.path.abspath(base_dir)}")
    else:
        print("Package 'vehicle' already exists.")

create_vehicle_package()
sys.path.append(os.getcwd())

try:
    from vehicle.types import cars, bikes
    print("Available Car Models:", cars.available_models())
    print("Available Bike Models:", bikes.available_models())
except ImportError as e:
    print(f"Error importing package: {e}")
```

Q2. Create a following DataFrame named as “data”. Write the python code for the following
    commands.

                            Company              Count            Price
             Pencil         Apsara               15               250
             Pencil         Natraj               20               200
             Pen            Cello                25               600
             Pen            Parkar               35               900
             Eraser         Apsara               20               300

           a) Find all rows with the label “Pencil”. Extract all columns
           b) Change the Eraser count as 25 instead of 20.
           c) List only the columns Company and Price.
           d) List only rows with labels ‘Pencil’ and ‘Pen’
           e) Add a new column named Colour to the DataFrame and display DataFrame..

```python
import pandas as pd

# Q2 OR: DataFrame operations
data = {
    'Company': ['Apsara', 'Natraj', 'Cello', 'Parkar', 'Apsara'],
    'Count': [15, 20, 25, 35, 20],
    'Price': [250, 200, 600, 900, 300]
}
labels = ['Pencil', 'Pencil', 'Pen', 'Pen', 'Eraser']
df = pd.DataFrame(data, index=labels)

print("Original DataFrame:")
print(df)

print("\n-- a) Rows with label 'Pencil' --")
print(df.loc['Pencil'])

print("\n-- b) Update Eraser Count --")
df.loc['Eraser', 'Count'] = 25
print(df.loc['Eraser'])

print("\n-- c) Columns Company and Price --")
print(df[['Company', 'Price']])

print("\n-- d) Rows 'Pencil' and 'Pen' --")
print(df.loc[['Pencil', 'Pen']])

print("\n-- e) Add 'Colour' Column --")
df['Colour'] = ['Black', 'Red', 'Blue', 'Green', 'White']
print(df)
```
