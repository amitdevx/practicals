Q2. Write a python program to check the validity of a password given by the user. The password should
    satisfy the following criteria:
           a) Contain at least 1 letter between a and z
           b) Contain at least 1 number between 0 and 9
           c) Contain at least 1 character from $, #, @
           d) Minimum length of password: 6
           e) Maximum length of password: 12
                                                                                              [20 Marks]

```python
# Read Iris.csv
if not os.path.exists("Iris.csv"):
    with open("Iris.csv", "w") as f:
        f.write("sepal_length,sepal_width,petal_length,petal_width,species\n")
        f.write("5.1,3.5,1.4,0.2,setosa\n")
        f.write("4.9,3.0,1.4,0.2,setosa\n")
        f.write("7.0,3.2,4.7,1.4,versicolor\n")
        f.write("6.4,3.2,4.5,1.5,versicolor\n")
        f.write("6.3,3.3,6.0,2.5,virginica\n")
    print("Created dummy Iris.csv")

df = pd.read_csv("Iris.csv")

print("DataFrame Contents:")
print(df)
print("\nStatistical Information:")
print(df.describe())

# Q2: Password validation
def validate_password(password):
    if len(password) < 6 or len(password) > 12:
        return False, "Password length must be between 6 and 12 characters."
    if not re.search(r"[a-z]", password):
        return False, "Must contain at least 1 lowercase letter."
    if not re.search(r"[0-9]", password):
        return False, "Must contain at least 1 digit."
    if not re.search(r"[$#@]", password):
        return False, "Must contain at least 1 special character ($, #, @)."
    return True, "Valid password."

try:
    pwd = input("Enter a password to validate: ")
    if not pwd: pwd = "Test@123"
except EOFError:
    pwd = "Test@123"

valid, msg = validate_password(pwd)
print(f"Password: {pwd}")
print(f"Result: {msg}")

# --- OR: File append + mobile validation ---
filename = "test_append.txt"
with open(filename, "w") as f:
    f.write("Original Content.\n")

try:
    fname = input("Enter filename (default: test_append.txt): ")
    if not fname: fname = filename

    string_to_append = input("Enter string to append: ")
    if not string_to_append: string_to_append = " Appended Content."

    if os.path.exists(fname):
        with open(fname, 'a') as f:
            f.write(string_to_append)
        print("\nUpdated File Content:")
        with open(fname, 'r') as f:
            print(f.read())
    else:
        print("File does not exist.")
except Exception as e:
    print(f"Error: {e}")

# Mobile number validation
def validate_mobile(number):
    pattern = r"^(0|91)?[6-9]\d{9}$"
    if re.match(pattern, number):
        return True
    return False

try:
    num = input("Enter mobile number: ")
    if not num: num = "9876543210"
except EOFError:
    num = "9876543210"

if validate_mobile(num):
    print("Valid Mobile Number")
else:
    print("Invalid Mobile Number")

if os.path.exists(filename):
    os.remove(filename)
```

Q2. Write a python program to list all files in a current directory. Display total count of .py files and
    content of any one .py files.

```python
# OR: List all files in current directory
current_dir = os.getcwd()
py_files = []

try:
    entries = os.listdir(current_dir)
    for entry in entries:
        full_path = os.path.join(current_dir, entry)
        if os.path.isfile(full_path):
            print(f"[FILE]   {entry}")
            if entry.endswith('.py'):
                py_files.append(entry)
        elif os.path.isdir(full_path):
            print(f"[FOLDER] {entry}")

    print(f"\nTotal .py files: {len(py_files)}")
    if py_files:
        print(f"\nContent of {py_files[0]}:")
        with open(os.path.join(current_dir, py_files[0]), 'r') as f:
            print(f.read())
except Exception as e:
    print(f"Error: {e}")
```
