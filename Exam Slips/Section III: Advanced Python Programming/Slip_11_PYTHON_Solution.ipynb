## Q1. Write a python code to a create a list of 10 integer numbers. Display square of all elements of list.
(Use a lambda function with map())
[10 Marks]

```python
numbers = [4, 7, 10, 13, 16, 19, 22, 25, 28, 31]
squares = list(map(lambda value: value ** 2, numbers))
print("Original numbers:", numbers)
print("Squares using map + lambda:", squares)
```

## Q2. Write a Python program that accepts a string from the user and performs the following tasks:
a) Split the string into individual words.
b) Identify and display all words that do not contain any vowels (a, e, i, o, u), ignore case while
checking for vowels.
c) Sort and display these words in alphabetical order.
d) Display the total count of such words.

```python
import re

text = "My crypt myths blend rhythm and sync with glyphs and sphynx figures."
words = [word.strip(",.!?;:") for word in text.split()]
words_without_vowels = sorted(
    {
        word
        for word in words
        if word and not re.search(r"[aeiou]", word, re.IGNORECASE)
    }
)
print("Input text:", text)
print("Words without vowels:", words_without_vowels)
print("Total words without vowels:", len(words_without_vowels))
```
