## Q1. Write a Python program to get the current time and demonstrate the following functions from the
time module:
a) Converts time in seconds to a struct_time in UTC
b) Converts a struct_time object into a formatted string.
[10 Marks]

```python
import time

current_seconds = time.time()
utc_struct = time.gmtime(current_seconds)
formatted_from_struct = time.strftime("%Y-%m-%d %H:%M:%S", utc_struct)
print("UTC struct_time:", utc_struct)
print("Formatted string from struct_time:", formatted_from_struct)
```

## Q2. Write a python program to check weather given url is valid or not? (for Ex:
https://www.yahoo.com)

```python
import re
from urllib.parse import urlparse

pattern = re.compile(r"^(https?://)([A-Za-z0-9-]+\.)+[A-Za-z]{2,}(/.*)?$")
urls = [
    "https://www.yahoo.com",
    "http://example.org",
    "https://invalid_domain",
    "ftp://legacy.server",
]

for url in urls:
    parsed = urlparse(url)
    is_valid = bool(pattern.match(url))
    print(f"{url} -> {'Valid' if is_valid else 'Invalid'}")
```
