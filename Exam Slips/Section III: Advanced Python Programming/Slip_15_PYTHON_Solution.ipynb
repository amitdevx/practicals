Q1. Write a Python program to get the current time and demonstrate the following functions from the
time module:
a) Converts time in seconds to a struct_time in UTC
b) Converts a struct_time object into a formatted string.
[10 Marks]

```python
import time

current_seconds = time.time()
utc_struct = time.gmtime(current_seconds)
formatted_from_struct = time.strftime("%Y-%m-%d %H:%M:%S", utc_struct)
print("UTC struct_time:", utc_struct)
print("Formatted string from struct_time:", formatted_from_struct)
```

Q2. Create a Web Page using Flask which contains following information about your Library Book List
a) Show the library name as a heading.
b) Display all available books in a table with title and author.
c) Add quote about reading in bold and blue color.
[20 Marks]

### OR

```python
from flask import Flask, render_template_string

app = Flask(__name__)

LIBRARY_NAME = "Visionary Science Library"
BOOKS = [
    {"title": "Clean Code", "author": "Robert C. Martin"},
    {"title": "Fluent Python", "author": "Luciano Ramalho"},
    {"title": "Python Tricks", "author": "Dan Bader"},
]

TEMPLATE = """
<!doctype html>
<title>{{ library }}</title>
<h1>{{ library }}</h1>
<table border=1 cellpadding=6>
  <tr><th>Title</th><th>Author</th></tr>
  {% for book in books %}
  <tr><td>{{ book.title }}</td><td>{{ book.author }}</td></tr>
  {% endfor %}
</table>
<p><strong style="color: blue;">"A reader lives a thousand lives before he dies."</strong></p>
"""

@app.route("/")
def home():
    return render_template_string(TEMPLATE, library=LIBRARY_NAME, books=BOOKS)

if __name__ == "__main__":
    app.run(debug=True, port=5002)
```

Q2. Write a python program to check weather given url is valid or not? (for Ex:
https://www.yahoo.com)

```python
import re
from urllib.parse import urlparse

pattern = re.compile(r"^(https?://)([A-Za-z0-9-]+\.)+[A-Za-z]{2,}(/.*)?$")
urls = [
    "https://www.yahoo.com",
    "http://example.org",
    "https://invalid_domain",
    "ftp://legacy.server",
]

for url in urls:
    parsed = urlparse(url)
    is_valid = bool(pattern.match(url))
    print(f"{url} -> {'Valid' if is_valid else 'Invalid'}")
```
