Q1. Write a Python program using regular expressions to validate whether a given string matches the
pattern of three-letter words such as bit, but, bat, hit, hut, hat.
[10 Marks]

```python
import re

pattern = r'^[bh][aiu]t$'

test_words = ['bit', 'but', 'bat', 'hit', 'hut', 'hat', 'cat', 'bet', 'hot', 'sit']

for word in test_words:
    if re.match(pattern, word):
        print(f"'{word}' matches the pattern.")
    else:
        print(f"'{word}' does NOT match the pattern.")
```

Q2. Consider a following record in DataFrame IPL.

| Player | Team | Category | BidPrice | Runs |
|---|---|---|---|---|
| Hardik Pandya | Mumbai Indians | Batsman | 13 | 1000 |
| K L Rahul | Kings Eleven | Batsman | 12 | 2400 |
| Andre Russel | Kolkata Knight Riders | Batsman | 7 | 900 |
| Jasprit Bumrah | Mumbai Indians | Bowler | 10 | 200 |
| Virat Kohli | RCB | Batsman | 17 | 3600 |
| Rohit Sharma | Mumbai Indians | Batsman | 15 | 3700 |

Create above DataFrame in python and write python code for following
a) Retrieve first 2 rows
b) Retrieve last 3 rows
c) Add null values in DataFrame.
d) Find most expensive player.
e) Print total players per team.
[20 Marks]

```python
import pandas as pd
import numpy as np

data = {
    'Player': ['Hardik Pandya', 'K L Rahul', 'Andre Russel', 'Jasprit Bumrah', 'Virat Kohli', 'Rohit Sharma'],
    'Team': ['Mumbai Indians', 'Kings Eleven', 'Kolkata Knight Riders', 'Mumbai Indians', 'RCB', 'Mumbai Indians'],
    'Category': ['Batsman', 'Batsman', 'Batsman', 'Bowler', 'Batsman', 'Batsman'],
    'BidPrice': [13, 12, 7, 10, 17, 15],
    'Runs': [1000, 2400, 900, 200, 3600, 3700]
}

df = pd.DataFrame(data)
print("IPL DataFrame:")
print(df)

# a) Retrieve first 2 rows
print("\na) First 2 rows:")
print(df.head(2))

# b) Retrieve last 3 rows
print("\nb) Last 3 rows:")
print(df.tail(3))

# c) Add null values in DataFrame
df.loc[1, 'BidPrice'] = np.nan
df.loc[4, 'Runs'] = np.nan
print("\nc) DataFrame after adding null values:")
print(df)

# d) Find most expensive player
most_expensive = df.loc[df['BidPrice'].idxmax()]
print(f"\nd) Most expensive player: {most_expensive['Player']} (BidPrice: {most_expensive['BidPrice']})")

# e) Print total players per team
print("\ne) Total players per team:")
print(df.groupby('Team')['Player'].count())
```

Q2. Create a package vehicle with a sub package types, sub package type contains modules cars and
bikes. Each module should have a function available_models() returning a list of models. Write a
python program to display all available models of cars and bikes.

```python
import os
import sys

base_dir = "vehicle"
types_dir = os.path.join(base_dir, "types")

os.makedirs(types_dir, exist_ok=True)

open(os.path.join(base_dir, "__init__.py"), 'w').close()
open(os.path.join(types_dir, "__init__.py"), 'w').close()

with open(os.path.join(types_dir, "cars.py"), "w") as f:
    f.write("def available_models():\n    return ['Sedan', 'SUV', 'Hatchback', 'Coupe']")

with open(os.path.join(types_dir, "bikes.py"), "w") as f:
    f.write("def available_models():\n    return ['Sport', 'Cruiser', 'Touring', 'Scooter']")

sys.path.insert(0, os.getcwd())

from vehicle.types import cars, bikes

print("Available Car Models:", cars.available_models())
print("Available Bike Models:", bikes.available_models())
```
