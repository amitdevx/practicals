Q1. Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical.
[10 Marks]


```python
try:
    num1 = input('Enter first number: ')
    num2 = input('Enter second number: ')
    num1 = float(num1)
    num2 = float(num2)
    print(f'Numbers: {num1}, {num2}')
except ValueError:
    raise TypeError('Inputs must be numerical values')
```

Q2. Create a DataFrame with Company, Count, Price columns and perform operations.
[20 Marks]
a) Find all rows with label 'Pencil'
b) Change Eraser count from 20 to 25
c) List Company and Price columns
d) List rows with 'Pencil' and 'Pen'
e) Delete Count column


```python
import pandas as pd
data = pd.DataFrame({
    'Company': ['Apsara', 'Natraj', 'Cello', 'Parker', 'Apsara'],
    'Type': ['Pencil', 'Pencil', 'Pen', 'Pen', 'Eraser'],
    'Count': [15, 20, 25, 35, 20],
    'Price': [250, 200, 600, 900, 300]
})
print('Original DataFrame:')
print(data)
print('\nPencil rows:')
print(data[data['Type'] == 'Pencil'])
data.loc[data['Type'] == 'Eraser', 'Count'] = 25
print('\nAfter updating Eraser count:')
print(data[['Company', 'Price']])
data_dropped = data.drop('Count', axis=1)
print('\nAfter deleting Count column:')
print(data_dropped)
```
