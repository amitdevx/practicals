Q1.Write a Python program to get the current time and demonstrate the following functions from the
time module:
a) Convert the current time in seconds to a readable string
b) Converts time in seconds to a struct_time in UTC
[10 Marks]

```python
import time

current_seconds = time.time()
human_readable = time.ctime(current_seconds)
utc_struct = time.gmtime(current_seconds)

print(f"Seconds since epoch: {current_seconds}")
print(f"Readable string via ctime(): {human_readable}")
print("UTC struct_time:", utc_struct)
```

Q2. Write a Python program that prompts the user to enter a text file name. If the file exists, open it in
read mode, display its contents line by line, and show the current file pointer position after each
line using ftell(). Then move the file pointer back to the beginning using seek() and display the file
content again. If the file does not exist, display an appropriate error message.
[20 Marks]

### OR

```python
from pathlib import Path

slip_dir = Path(__file__).resolve().parent if '__file__' in globals() else Path.cwd()
sample_file = slip_dir / "story_notes.txt"
sample_text = """Chapter 1
Insightful line number two.
Another observation closes the chapter.
"""
sample_file.write_text(sample_text, encoding="utf-8")

filename = sample_file.name
print(f"Reading file: {filename}")
try:
    with sample_file.open("r", encoding="utf-8") as handle:
        for line in handle:
            print(line.rstrip())
            print(f"Pointer after line: {handle.tell()} bytes")
        handle.seek(0)
        print("\nRe-reading entire file after seek(0):\n" + handle.read())
except FileNotFoundError:
    print(f"File {filename} not found.")
```

```python
import pandas as pd

student1 = pd.DataFrame([
    {"ID": "S1", "Name": "Anita", "Age": 18},
    {"ID": "S2", "Name": "Sakshi", "Age": 23},
    {"ID": "S3", "Name": "Om", "Age": 20},
    {"ID": "S4", "Name": "Raj", "Age": 21},
    {"ID": "S5", "Name": "Kanika", "Age": 20},
])

student2 = pd.DataFrame([
    {"ID": "S2", "City": "Pune", "Marks": 97},
    {"ID": "S4", "City": "Nashik", "Marks": 78},
    {"ID": "S5", "City": "Mumbai", "Marks": 86},
    {"ID": "S6", "City": "Delhi", "Marks": 66},
    {"ID": "S7", "City": "Nanded", "Marks": 54},
])

print("Student 1 DataFrame:\n", student1, sep="")
print("\nStudent 2 DataFrame:\n", student2, sep="")

merged = pd.merge(student1, student2, on="ID", how="inner")
print("\nMerged DataFrame (matching records from both sides):\n", merged, sep="")
```
