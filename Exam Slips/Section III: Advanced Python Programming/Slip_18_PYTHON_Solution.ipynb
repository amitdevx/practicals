Q2. Write a Python program to validate mobile number based on following criteria
      a) The first digit should contain numbers between 6 to 9.
      b) The rest 9 digit can contain any number between 0 to 9.
      c) The mobile number can have 11 digits also by including 0 at the starting.
      d) The mobile number can be of 12 digits also by including 91 at the starting
                                                                                            [20 Marks]

### OR

```python
import re

pattern = re.compile(r"^(0|91)?[6-9]\d{9}$")
mobile_numbers = [
    "9823456789",
    "09823456789",
    "919823456789",
    "5123456789",
]

for number in mobile_numbers:
    is_valid = bool(pattern.match(number))
    print(f"{number} -> {'Valid' if is_valid else 'Invalid'}")
```

Q2. Build a package education with a sub package courses, containing modules science.py and
    arts.py. Each module should have a function get_subjects(). Write a python program to display all
    subjects.

```python
from pathlib import Path
import importlib.util
import sys

slip_dir = Path(__file__).resolve().parent if '__file__' in globals() else Path.cwd()
courses_dir = slip_dir / "education" / "courses"
courses_dir.mkdir(parents=True, exist_ok=True)
(courses_dir / "__init__.py").write_text("", encoding="utf-8")
(courses_dir.parent / "__init__.py").write_text("", encoding="utf-8")

(courses_dir / "science.py").write_text("""def get_subjects():
    return ["Physics", "Chemistry", "Biology"]
""", encoding="utf-8")
(courses_dir / "arts.py").write_text("""def get_subjects():
    return ["History", "Political Science", "Psychology"]
""", encoding="utf-8")

sys.path.insert(0, str(slip_dir))
science_spec = importlib.util.spec_from_file_location("education.courses.science", courses_dir / "science.py")
science = importlib.util.module_from_spec(science_spec)
science_spec.loader.exec_module(science)
arts_spec = importlib.util.spec_from_file_location("education.courses.arts", courses_dir / "arts.py")
arts = importlib.util.module_from_spec(arts_spec)
arts_spec.loader.exec_module(arts)

print("Science subjects:", science.get_subjects())
print("Arts subjects:", arts.get_subjects())
```
