**Q1.** Write a python program to accept a string, extract all words that start with a capital letter and remove all special characters.


```python
import re

s = input("Enter a string: ")
if not s: s = "Hello World! This is Python."

words = s.split()
capital_words = [w for w in words if w and w[0].isupper()]
print(f"Words starting with capital: {capital_words}")

clean_s = re.sub(r'[^a-zA-Z0-9\s]', '', s)
print(f"String without special chars: {clean_s}")
```

**Q2 (OR Option).** Write a python program to accept file name. If file exist copy contents...


```python
filename = input("Enter source file name: ")
dest_filename = input("Enter destination file name: ")

if not filename: 
    with open("test_src.txt", "w") as f: f.write("This is a test file.\nWith multiple lines.")
    filename = "test_src.txt"
    dest_filename = "test_dest.txt"

try:
    with open(filename, 'r') as f1:
        content = f1.read()
    
    with open(dest_filename, 'w') as f2:
        f2.write(content)
    
    print(f"Content copied to {dest_filename}.")
    print("New File Content:")
    print(content)
    
    lines = len(content.splitlines())
    words = len(content.split())
    chars = len(content)
    print(f"Total Lines: {lines}")
    print(f"Total Words: {words}")
    print(f"Total Characters: {chars}")

except FileNotFoundError:
    print("File not found.")
```
