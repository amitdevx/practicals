### 1.

Represent the following information using a bar graph (in brown color ) ,

| Subject | Marks |
| --- | --- |

### 4.

Using sympy declare points P (7,2) , Q (1,8). Find the distance between them. Declare the line
passing through them. Find the slope and equation of the line passing through points P and Q.


```python
import sympy as sp
P = sp.Point(7, 2)

Q = sp.Point(1, 8)

dist = P.distance(Q)

line = sp.Line(P, Q)

print(f"Distance: {dist}")

print(f"Slope: {line.slope}")

print(f"Equation: {line.equation()} = 0")
```

### 5.

Using sympy declare  ABC where A (2 , 2), B (4, 2), C ( 3, 6) ,for this triangle find its centroid
and also translate  ABC in x direction by 2 units.


```python
import sympy as sp
A, B, C = sp.Point(2, 2), sp.Point(4, 2), sp.Point(3, 6)

tri = sp.Triangle(A, B, C)

print(f"Centroid: {tri.centroid}")

trans_verts = [p.translate(2, 0) for p in tri.vertices]

print("Translated Vertices (x+2):", trans_verts)
```

### Q2.1 (Attempt Any One)

1. Write a python program to plot the polygon with vertices at (1, 3), (2, 4), (4 , 2) and ( 3,1)
Plot its rotation about ,origin through angles 1)
𝜋
2
radians and 2) π radians .
Also plot x and y axis in black colours.

```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
def apply_transformation(points, matrix):
    return np.dot(points, matrix.T)

poly = np.array([[1, 3], [2, 4], [4, 2], [3, 1]])

poly_plot = np.vstack([poly, poly[0]])

Rot1 = np.array([[0, -1], [1, 0]])

Rot2 = np.diag([-1, -1])

p1 = apply_transformation(poly, Rot1)

p1_plot = np.vstack([p1, p1[0]])

p2 = apply_transformation(poly, Rot2)

p2_plot = np.vstack([p2, p2[0]])

plt.figure()

plt.axhline(0, color='black')

plt.axvline(0, color='black')

plt.plot(poly_plot[:, 0], poly_plot[:, 1], 'b-', label='Original')

plt.plot(p1_plot[:, 0], p1_plot[:, 1], 'g--', label='Rot pi/2')

plt.plot(p2_plot[:, 0], p2_plot[:, 1], 'r-.', label='Rot pi')

plt.legend()

plt.grid(True)

plt.axis('equal')

plt.title("Polygon Rotations")

plt.show()
```

2. Write a Python program to generate and plot 50 points on parabola y2 = 4x with 4 ≤ 𝑥 ≤ 16
using iterative formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
k = 4 # from y^2 = 4x
num_points = 50
x_min, x_max = 0, 10

# Generate points
x = np.linspace(x_min, x_max, num_points)
y_pos = np.sqrt(k * x)
y_neg = -np.sqrt(k * x)

# Plot
plt.figure()
plt.plot(x, y_pos, 'b-', label=f'y^2 = {k}x')
plt.plot(x, y_neg, 'b-')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Parabola: y^2 = {k}x')
plt.grid(True)
plt.legend()
plt.show()
```
