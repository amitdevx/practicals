### 1.

Represent the following information using pie chart with detached sectors .

| Animal Type | Number |
| --- | --- |
| Wild Animals | 150 |
| Grazing Animals | 400 |
| Birds | 225 |
| Water Animals | 175 |
| Reptiles | 50 |



```python
import matplotlib.pyplot as plt

categories = ['Wild Animals', 'Grazing Animals', 'Birds', 'Water Animals', 'Reptiles']
counts = [150, 400, 225, 175, 50]
explode = [0.05] * len(categories)

fig, ax = plt.subplots(figsize=(6.5, 6.5))
ax.pie(counts, labels=categories, autopct='%1.1f%%', explode=explode,
       startangle=90, shadow=True)
ax.set_title('Distribution of Animals (Detached Sectors)')
ax.axis('equal')
plt.show()
```

### 3.

Plot Surface plot for the function f(x, y) = –x3 – y3, where x, y  [1, 10].


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

x = np.linspace(1, 10, 100)
y = np.linspace(1, 10, 100)
X, Y = np.meshgrid(x, y)
Z = -(X**3 + Y**3)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma', edgecolor='none')
ax.set_title(r'Surface Plot of $f(x, y) = -(x^3 + y^3)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```

### 4.

Using sympy find the angle between the line segment AB joining the points A[7,2], B[3,−4]
and the line having equation 2x − y = 1 .


```python
from sympy import Point, Line, acos

A = Point(7, 2)
B = Point(3, -4)
segment_vector = B - A

given_line = Line(Point(0, 0.5), Point(0.5, 1))  # 2x - y = 1
line_vector = given_line.direction

dot_product = segment_vector.dot(line_vector)
angle = acos(dot_product / (segment_vector.distance(Point(0, 0)) * line_vector.distance(Point(0, 0))))

print(f'Angle between segment AB and line 2x - y = 1 (radians): {angle}')
```

### 5.

Using sympy declare polygon ABCD with vertices A(3, 3), B(3, 6), C( 8, 6) and D(7, 2) ,for
this polygon determine whether point (4,4) inside . And scale ABCD in x and y directions
by factors 4 and 3 respectively.


```python
from sympy import Point, Polygon

A = Point(3, 3)
B = Point(3, 6)
C = Point(8, 6)
D = Point(7, 2)
polygon = Polygon(A, B, C, D)
test_point = Point(4, 4)

print(f'Is point (4, 4) inside polygon ABCD? {polygon.encloses_point(test_point)}')

scaled_points = [Point(pt.x * 4, pt.y * 3) for pt in (A, B, C, D)]
for label, pt in zip(["A'", "B'", "C'", "D'"], scaled_points):
    print(f'Scaled point {label}: {pt}')
```

Write a python program to find the transformation of the line segment joining the points
A(1,−2) and B(3,5) for the following sequence of transformations:
(I) Rotation about origin through an angle π.

(II) Scaling in X− coordinate by 7 units.

(III) Uniform scaling by factor 4. Plot original and transformed line segment.

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([1.0, -2.0])
B = np.array([3.0, 5.0])

rotation = np.array([[-1, 0], [0, -1]], dtype=float)
scale_x = np.array([[7, 0], [0, 1]], dtype=float)
uniform_scale = np.array([[4, 0], [0, 4]], dtype=float)

transform_matrix = uniform_scale @ scale_x @ rotation

A_trans = transform_matrix @ A
B_trans = transform_matrix @ B

plt.figure(figsize=(6, 6))
plt.plot([A[0], B[0]], [A[1], B[1]], marker='o', color='navy', label='Original AB')
plt.plot([A_trans[0], B_trans[0]], [A_trans[1], B_trans[1]], marker='s', color='crimson', label="Transformed A'B'")

plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.title('Transformations of Line Segment AB')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle=':')
plt.show()
```

2.Write python program to generate and plot 50 points on parabola y2 = 28 x , using iterative
formulae for 0  x  100.

–––––––––––––––

*

–––––––––––––––

```python
import matplotlib.pyplot as plt

num_points = 50
max_x = 100
step = max_x / (num_points - 1)

xs = []
ys = []
x_val = 0.0
for _ in range(num_points):
    y_val = (28 * x_val) ** 0.5
    xs.append(x_val)
    ys.append(y_val)
    x_val += step

plt.figure(figsize=(7, 4))
plt.plot(xs, ys, 'o', color='darkgreen', label=r'$y^2 = 28x$')
plt.title(r'Points on Parabola $y^2 = 28x$ for $0 \leq x \leq 100$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```
