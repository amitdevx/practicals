### 1.

Represent the following information using a horizontal bar graph (in brown color ) ,

| Subject | Marks |
| --- | --- |
| Mathematics | 45 |
| English | 35 |
| Biology | 85 |
| Physics | 50 |
| Chemistry | 60 |



```python
import matplotlib.pyplot as plt

subjects = ["Mathematics", "English", "Biology", "Physics", "Chemistry"]
marks = [45, 35, 85, 50, 60]

plt.figure(figsize=(8, 5))
plt.barh(subjects, marks, color="saddlebrown")
plt.title("Horizontal Bar Graph of Marks")
plt.xlabel("Marks")
plt.ylabel("Subject")
plt.xlim(0, 100)
plt.grid(axis="x", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

### 2.

Plot the graph of 𝑦 = 𝑥 3 − 3𝑥 + 2, for x  [–5, 5] in maroon colour with arrow lines.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 400)
y = x**3 - 3 * x + 2

plt.figure(figsize=(8, 5))
plt.plot(x, y, color="maroon", linewidth=2, marker=">", markevery=20)
plt.title(r"Graph of $y = x^3 - 3x + 2$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, linestyle=":", alpha=0.5)
plt.show()
```

### 3.

Draw surface plot for the function 𝑓(𝑥, 𝑦) = √𝑥 2 + 𝑦 2 when –20 x, y 20,
with red colour.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

grid = np.linspace(-20, 20, 150)
X, Y = np.meshgrid(grid, grid)
Z = np.sqrt(X**2 + Y**2)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, color="red", edgecolor="none", alpha=0.9)
ax.set_title(r"Surface Plot of $f(x, y) = \sqrt{x^2 + y^2}$")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
plt.show()
```

### 4.

Using sympy declare points L(–5,0) , M(4,0) and N(9,0). Determine whether the points L, M
and N are collinear. Declare the line passing through the point L and N, find the slope and
equation of the line passing through points L and N.


```python
from sympy import Point, Line

L = Point(-5, 0)
M = Point(4, 0)
N = Point(9, 0)

collinear = Point.is_collinear(L, M, N)
line_ln = Line(L, N)

print(f"Points L{L}, M{M}, N{N}")
print(f"Are L, M, N collinear? {'Yes' if collinear else 'No'}")
print(f"Slope of line through L and N: {line_ln.slope}")
print(f"Equation of line LN: {line_ln.equation()}")
```

### 5.

Using sympy declare regular polygon with center at origin, radius 2 and number of sides 5. Find
its area, perimeter, centroid and also translate it in y direction by 4 units.


```python
from sympy import Point, RegularPolygon

center = Point(0, 0)
polygon = RegularPolygon(center, 2, 5)

area = polygon.area
perimeter = polygon.perimeter
centroid = polygon.centroid
translated_polygon = polygon.translate(0, 4)

print(f"Regular pentagon vertices: {polygon.vertices}")
print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
print(f"Centroid: {centroid}")
print(f"Translated vertices (shifted +4 along y): {translated_polygon.vertices}")
```

### Q2.1 (Attempt Any One)

1. Write a python program to plot the polygon with vertices at points A(0,0), B(1, 0), C(1, 1)
and D (0 , 1), find the transformed polygon by applying combined transformations using
sequence a) reflections thorugh x-axis,
b) rotation about origin through angle 450
and
c) uniform scaling by factor -3 units. Both the figures must be in different colours.

```python
import numpy as np
import matplotlib.pyplot as plt

vertices = np.array([[0, 0], [1, 0], [1, 1], [0, 1]])

def close_shape(points):
    return np.vstack([points, points[0]])

original = close_shape(vertices)

reflection_matrix = np.array([[1, 0], [0, -1]])
rotation_matrix = np.array(
    [[np.cos(np.deg2rad(45)), -np.sin(np.deg2rad(45))], [np.sin(np.deg2rad(45)), np.cos(np.deg2rad(45))]]
)
scaling_matrix = -3 * np.eye(2)

transformed_vertices = (reflection_matrix @ vertices.T).T
transformed_vertices = (rotation_matrix @ transformed_vertices.T).T
transformed_vertices = (scaling_matrix @ transformed_vertices.T).T
transformed = close_shape(transformed_vertices)

fig, ax = plt.subplots(figsize=(7, 7))
ax.plot(original[:, 0], original[:, 1], color="blue", linewidth=2, label="Original")
ax.plot(transformed[:, 0], transformed[:, 1], color="orange", linewidth=2, label="Transformed")
ax.axhline(0, color="black", linewidth=0.5)
ax.axvline(0, color="black", linewidth=0.5)
ax.set_aspect("equal")
ax.set_title("Combined Transformations on Square")
ax.legend()
ax.grid(True, linestyle=":")
plt.show()
```

2. Write a python program to generate and plot 50 points on parabola 𝑦 2 = 12𝑥 , in the where
0 ≤ 𝑥 ≤ 20 using iterative formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

num_points = 50
t_min = -np.sqrt(12 * 20)
t_max = np.sqrt(12 * 20)
t_values = np.linspace(t_min, t_max, num_points)

points = np.array([(t**2 / 12, t) for t in t_values if 0 <= t**2 / 12 <= 20])

plt.figure(figsize=(7, 5))
plt.scatter(points[:, 0], points[:, 1], color="crimson", label=r"$y^2 = 12x$")
plt.title("Iterative Points on Parabola")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0, 20)
plt.grid(True, linestyle=":")
plt.legend()
plt.show()
```
