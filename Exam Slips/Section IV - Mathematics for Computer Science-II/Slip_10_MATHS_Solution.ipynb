### 1.

Represent the following information using a horizontal bar graph (in blue color ) ,

| Year | Number of Matches |
| --- | --- |
| 2020 | 50 |
| 2021 | 40 |
| 2022 | 30 |
| 2023 | 67 |
| 2024 | 45 |



```python
import matplotlib.pyplot as plt
years = ['2020', '2021', '2022', '2023', '2024']

matches = [50, 40, 30, 67, 45]

plt.figure()

plt.barh(years, matches, color='blue')

plt.xlabel('Number of Matches')

plt.ylabel('Year')

plt.title('Matches per Year')

plt.show()
```

### 2.

Plot the graph of y = x2 –5 , for x  [–10, 20] in red colour with lines using arrow marker.


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 20, 20)

y = x**2 - 5

plt.figure()

plt.plot(x, y, 'r-', marker='>', label='y = x^2 - 5')

plt.xlabel('x')

plt.ylabel('y')

plt.legend()

plt.grid(True)

plt.show()
```

### 3.

Draw Wireframe plot for the function 𝑓(𝑥, 𝑦) = 𝑐𝑜𝑠 √𝑥 2 − 𝑦 2 when 0 x, y 2𝜋,
with green colour.


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2*np.pi, 50)

y = np.linspace(0, 2*np.pi, 50)

X, Y = np.meshgrid(x, y)

with np.errstate(invalid='ignore'):
    Z = np.cos(np.sqrt(X**2 - Y**2))

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(X, Y, Z, color='green')

ax.set_title('Wireframe Plot')

plt.show()
```

### Q2.1 (Attempt Any One)

1. Write a python program to plot the triangle OAB where O (0, 0), A(8, –9), B( 7, 4) ,for this
triangle find a) scaling in x and y co-ordinate by factor 5 and–6 respectively, b) reflection
through x axis and c) rotation about origin through angle 600 . And plot all transformed
figures in different colours.

```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
def apply_transformation(points, matrix):
    return np.dot(points, matrix.T)

poly = np.array([[0, 0], [8, -9], [7, 4]])

poly_plot = np.vstack([poly, poly[0]])

S = np.diag([5, -6])

Ref = np.diag([1, -1])

theta = np.radians(60)

c, s = np.cos(theta), np.sin(theta)

Rot = np.array([[c, -s], [s, c]])

p1 = apply_transformation(poly, S)

p1_plot = np.vstack([p1, p1[0]])

p2 = apply_transformation(p1, Ref)

p2_plot = np.vstack([p2, p2[0]])

p3 = apply_transformation(p2, Rot)

p3_plot = np.vstack([p3, p3[0]])

plt.figure()

plt.plot(poly_plot[:, 0], poly_plot[:, 1], 'b-', label='Original')

plt.plot(p1_plot[:, 0], p1_plot[:, 1], 'g--', label='Scaled')

plt.plot(p2_plot[:, 0], p2_plot[:, 1], 'c--', label='Reflected')

plt.plot(p3_plot[:, 0], p3_plot[:, 1], 'r-', label='Rotated (Final)')

plt.legend()

plt.grid(True)

plt.axis('equal')

plt.title("Triangle Transformations")

plt.show()
```

### Q2.2

2. Write a python program to generate and plot100 points on circle 𝑥 2 + 𝑦 2 = 144 using
iterative formulae.

```python
import numpy as np
import matplotlib.pyplot as plt

r = 12.0
n_points = 100

theta = np.linspace(0, 2*np.pi, n_points)

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure()
plt.plot(x, y, label=f'x^2 + y^2 = 144')
plt.title(f'Circle x^2 + y^2 = 144 (Iterative)')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal') # Ensure circle looks circular
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
