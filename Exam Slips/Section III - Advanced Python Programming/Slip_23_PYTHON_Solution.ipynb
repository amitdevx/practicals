## Q1. Write a Python program that defines a function productInfo() which accepts a product name and its
    price. The function should display the product name only if its price is less than Rs.100. If no price
    is provided, the function should assume the default price to be Rs.100.
                                                                                            [10 Marks]


```python
def productInfo(name, price=100):
    if price < 100:
        print(f"Product Name: {name}, Price: {price}")
    else:
        pass

productInfo("Pen", 20)
productInfo("Notebook", 150)
productInfo("Eraser")
```

## Q2. Write a Python program that:
    a) Retrieves current local time and UTC time.
    b) Prints the current time using ctime() and gmtime().
    c) Accepts a date-time string input in the format DD-MM-YYYY HH:MM:SS.
    d) Parses the string using strptime() and formats it back using strftime() and display it in two custom
        formats.
                                                                                            [20 Marks]

```python
import os
from datetime import datetime

# Q2: Show date-time info and directory listing
try:
    dt_str = input("Enter date-time (DD-MM-YYYY HH:MM:SS): ")
    dt = datetime.strptime(dt_str, "%d-%m-%Y %H:%M:%S")
except Exception:
    dt = datetime.now()
    print(f"Using current date-time: {dt.strftime('%d-%m-%Y %H:%M:%S')}")

print(f"Day of week: {dt.strftime('%A')}")
print(f"Day of month: {dt.day}")
print(f"Day of year: {dt.timetuple().tm_yday}")

print(f"\nFiles in current directory:")
for item in os.listdir('.'):
    print(f"  {item}")
```

## Q2. Write a Python program that accepts a string from the user and performs the following tasks:
   a) Split the string into individual words.
   b) Identify and display all words that do not contain any vowels (a, e, i, o, u), ignore case while
       checking for vowels.
   c) Sort and display these words in alphabetical order.
   d) Display the total count of such words.

```python
try:
    text = input("Enter a string: ")
except Exception:
    text = "Sky Fly Cry Apple Banana Rhythm"
    print(f"Input: {text}")

words = text.split()
vowels = set("aeiou")
no_vowel_words = []

for word in words:
    if not any(char.lower() in vowels for char in word):
        no_vowel_words.append(word)

print("Words with no vowels:", no_vowel_words)
no_vowel_words.sort()
print("Sorted words with no vowels:", no_vowel_words)
print(f"Total count: {len(no_vowel_words)}")
```
