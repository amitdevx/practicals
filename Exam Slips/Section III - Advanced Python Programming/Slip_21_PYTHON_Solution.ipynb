## Q1. Create a DataFrame from a dictionary {Subject_Name : Marks} and display the first 5 rows and
    last 5 rows of DataFrame.
                                                                                               [10 Marks]

```python
import pandas as pd

data = {
    "Mathematics": 89,
    "Statistics": 92,
    "Python": 95,
    "Data Science": 88,
    "Operating Systems": 84,
    "Networking": 81,
    "Machine Learning": 90,
}

df = pd.DataFrame([
    {"Subject_Name": subject, "Marks": marks}
    for subject, marks in data.items()
])
print("First five rows:\n", df.head(), sep="")
print("\nLast five rows:\n", df.tail(), sep="")
```

## Q2. Write a Python program that accepts a filename from the user, opens the file in read mode and
    display the contents of files. The program should compute and display the total number of
    characters (including spaces), words, and lines present in the file. If file does not exist give
    appropriate message.
                                                                                               [20 Marks]

```python
from pathlib import Path

slip_dir = Path(__file__).resolve().parent if '__file__' in globals() else Path.cwd()
file_path = slip_dir / "sample_notes.txt"
file_path.write_text("""Sem IV Advanced Python practicals.
Focus on pandas, flask, and regex modules.
Practice every day.
""", encoding="utf-8")

with file_path.open("r", encoding="utf-8") as handle:
    content = handle.read()

characters = len(content)
words = len(content.split())
lines = len(content.splitlines())

print(f"File: {file_path}")
print(f"Characters (with spaces): {characters}")
print(f"Words: {words}")
print(f"Lines: {lines}")
print("\nContent:\n" + content)
```

## Q2. Create a package banking with modules account.py and loan.py.
    a) account.py should contain function open_account(name) that accepts a user’s name and prints a
    message confirming account creation.
    b) loan.py should contain function apply_loan(amount) that accepts a loan amount and prints a confirmation
    message with the amount.
    Write a python program to import modules from banking package and demonstrate both functions.

```python
from pathlib import Path
import importlib.util
import sys

slip_dir = Path(__file__).resolve().parent if '__file__' in globals() else Path.cwd()
banking_dir = slip_dir / "banking"
banking_dir.mkdir(exist_ok=True)
(banking_dir / "__init__.py").write_text("", encoding="utf-8")

(banking_dir / "account.py").write_text("""def open_account(name):
    return f"Account created successfully for {name}."
""", encoding="utf-8")

(banking_dir / "loan.py").write_text("""def apply_loan(amount):
    return f"Loan application received for amount: Rs.{amount}."
""", encoding="utf-8")

sys.path.insert(0, str(slip_dir))
account_spec = importlib.util.spec_from_file_location("banking.account", banking_dir / "account.py")
account = importlib.util.module_from_spec(account_spec)
account_spec.loader.exec_module(account)
loan_spec = importlib.util.spec_from_file_location("banking.loan", banking_dir / "loan.py")
loan = importlib.util.module_from_spec(loan_spec)
loan_spec.loader.exec_module(loan)

print(account.open_account("Anita Kulkarni"))
print(loan.apply_loan(250000))
```
