<h2>Q1. Plot the graph of algebraic functions</h2>

<h2>(a) Plot the graph of $f(x) = x$ </h2>

```python
from pylab import *

x = linspace(-5, 5, 100)
y = x**3

plot(x, y)
xlabel('x')
ylabel('y')
title('Graph of f(x) = x^3')
grid(True)
show()
```

<h2>(b) Plot the graph of $f(x) = 1 + x + 2x^2 + 3x^3 + 4x^4$
</h2>

```python
from pylab import*
x=linspace (-10,10,100)
y=1+x+2*x**2+3*x**3+4*x**4
plot(x,y)
xlabel('x')
ylabel('y')
title('Graph ')
show()
```

<h2>Q2. Plot the graph of f$f(x) = \sin x$
 and $g(x) = \cos x$
</h2>

```python
from pylab import *

x = linspace(-2*pi, 2*pi, 100)
f = sin(x)
g = cos(x)

plot(x, f, label='sin x')
plot(x, g, label='cos x')
xlabel('x')
ylabel('y')
title('Graph of sin(x) and cos(x)')
legend()
grid(True)
show()
```

<h2>Q3. Plot the graph of the exponential function
$f(x) = e^{-x^2}$
2</h2>

```python
from pylab import *

x = linspace(-5, 5, 100)
y = exp(-x**2)

plot(x, y, 'g')
xlabel('x')
ylabel('y')
title('Graph of f(x) = e^(-x^2)')
grid(True)
show()
```

<h2>Q4. Draw a bar graph for monthly expenditure</h2>


| Item     | Expenditure |
| -------- | ----------- |
| Clothing | 600         |
| Food     | 4000        |
| Rent     | 2000        |
| Petrol   | 1500        |
| Mess     | 700         |

```python
from pylab import *

items = ['Clothing', 'Food', 'Rent', 'Petrol', 'Mess']
expenditure = [600, 4000, 2000, 1500, 700]

bar(items, expenditure, color='green')
xlabel('Items')
ylabel('Expenditure')
title('Monthly Expenditure')
grid(axis='y')
show()
```

<h2>Q5. Draw a bar graph showing student performance</h2>

| Subject | Percentage |
| ------- | ---------- |
| Maths   | 60         |
| Science | 90         |
| English | 70         |
| Marathi | 85         |
| Hindi   | 91         |

```python
from pylab import *

subjects = ['Maths', 'Science', 'English', 'Marathi', 'Hindi']
percentage = [60, 90, 70, 85, 91]

bar(subjects, percentage, color='pink')
xlabel('Subjects')
ylabel('Percentage')
title('Student Performance by Subject')
grid(axis='y')
show()
```
