# Uniform Points on Parabolas

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Q1. Generate and plot n uniformly spaced points on the parabola y=x² for x ∈[−3,3]

```python
import numpy as np
import matplotlib.pyplot as plt
n = 10
x = np.linspace(-3, 3, n)
y = x**2
plt.plot(x, y, 'o-')
plt.title("Uniform Points on y = x² (x-range)")
plt.grid()
plt.show()
```

## Q2. Generate and plot uniformly spaced points on the parabola y=2x² for x ∈ [−2,2]

```python
import numpy as np
import matplotlib.pyplot as plt
n=8
x = np.linspace(-2, 2, n)
y = 2 * x**2
plt.plot(x, y, 'o')
plt.title("Uniform Points on y = 2x²")
plt.grid()
plt.show()
```

## Q4. Generate uniformly spaced points on the parabola y = (x−1)²+2 for x ∈ [−2,4]

## Q6. Generate and plot uniformly spaced points on the parabola y=x² for y ∈ [0,9]

```python
import numpy as np
import matplotlib.pyplot as plt
n = 10
y = np.linspace(0, 9, n)
x = np.sqrt(y)
plt.plot(x, y, 'o-')
plt.plot(-x, y, 'o-')
plt.title("Uniform Points using y-range on y = x²")
plt.grid()
plt.show()
```

## Q7. Generate and plot uniformly spaced points on the parabola y=x² using both x-range and y-range methods.

```python
import numpy as np
import matplotlib.pyplot as plt
n = 10
x1 = np.linspace(-3, 3, n)
y1 = x1**2
y2 = np.linspace(0, 9, n)
x2 = np.sqrt(y2)
plt.plot(x1, y1, 'o-', label="x-range")
plt.plot(x2, y2, 's-', label="y-range")
plt.plot(-x2, y2, 's-')
plt.legend()
plt.title("Parabola: x-range vs y-range")
plt.grid()
plt.show()
```
