# Title: Python Functions

---
## 1. User-Defined Function

**Definition:** A user-defined function is a function created by the programmer to perform a specific task. It helps in code reusability, readability, and modular programming.

**Syntax:**
```python
def function_name(parameters):
    statements
```

````python
def add_numbers(a, b):
    result = a + b
    print("Sum =", result)

add_numbers(10, 20)
````

---
## 2. Calling a Function

**Explanation:** Calling a function means executing the function by using its name followed by parentheses ().

````python
def greet():
    print("Welcome to Python Programming")

# Function call
greet()
````

---
## 3. Function Arguments

**Explanation:** Arguments are values passed to a function when it is called. They allow functions to work with different data.

### Types of Arguments:
1. **Positional arguments** - Arguments passed in the order they are defined
2. **Keyword arguments** - Arguments passed with parameter names
3. **Default arguments** - Arguments with default values

````python
def student_info(name, age):
    print("Name:", name)
    print("Age:", age)

# Positional arguments
student_info("Amit", 20)

# Keyword arguments
student_info(age=21, name="Neha")
````

---
## 4. Return Statement

**Explanation:** The return statement is used to send a value back from the function to the caller.

````python
def square(n):
    return n * n

result = square(5)
print("Square =", result)
````

---
## 5. Function Returning Multiple Values

**Explanation:** A function can return more than one value using tuples.

````python
def arithmetic_operations(a, b):
    add = a + b
    sub = a - b
    mul = a * b
    return add, sub, mul

x, y, z = arithmetic_operations(10, 5)
print("Addition =", x)
print("Subtraction =", y)
print("Multiplication =", z)
````

---
## 6. Anonymous Function (Lambda Function)

**Explanation:** An anonymous function is a function without a name. It is created using the lambda keyword and is generally used for short operations.

**Syntax:**
```python
lambda arguments: expression
```

````python
square = lambda x: x * x
print("Square =", square(6))
````

````python
add = lambda a, b: a + b
print("Sum =", add(4, 7))
````

---
## 7. Recursive Function

**Explanation:** A recursive function is a function that calls itself to solve a problem. It must have a base condition to stop recursion.

### Example 1: Factorial

````python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print("Factorial =", factorial(5))
````

### Example 2: Fibonacci Series

````python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

for i in range(6):
    print(fibonacci(i), end=" ")
````
