# Exception Handling - Set A

## Q1. Write a Python program that prompts the user to input a number to handle a ZeroDivisionError exception when dividing a number by zero.

```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    result = num1 / num2
    print(f"Result: {result}")
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")
```

## Q2. Write a Python program that prompts the user to input an integer and raises a ValueError exception if the input is not a valid integer.

```python
try:
    num = int(input("Enter an integer: "))
    print(f"You entered: {num}")
except ValueError:
    print("Error: Invalid input! Please enter a valid integer.")
```

## Q3. Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical.

```python
try:
    num1 = input("Enter first number: ")
    num2 = input("Enter second number: ")
    
    if not num1.replace('.', '', 1).lstrip('-').isdigit():
        raise TypeError("First input is not numerical")
    if not num2.replace('.', '', 1).lstrip('-').isdigit():
        raise TypeError("Second input is not numerical")
    
    result = float(num1) + float(num2)
    print(f"Sum: {result}")
except TypeError as e:
    print(f"Error: {e}")
```

## Q4. Write a Python program that executes an operation on a list and handles an IndexError exception if the index is out of range.

```python
try:
    my_list = [10, 20, 30, 40, 50]
    print(f"List: {my_list}")
    
    index = int(input("Enter index to access: "))
    print(f"Element at index {index}: {my_list[index]}")
except IndexError:
    print("Error: Index out of range!")
```
