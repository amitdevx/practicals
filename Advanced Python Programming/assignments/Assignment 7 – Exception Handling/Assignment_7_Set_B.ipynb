# Exception Handling - Set B

## Q1. Write a Python program that prompts the user to input a number and handles a KeyboardInterrupt exception if the user cancels the input.

```python
try:
    num = int(input("Enter a number (Press Ctrl+C to cancel): "))
    print(f"You entered: {num}")
except KeyboardInterrupt:
    print("\nInput cancelled by user!")
```

## Q2. Write a Python program that executes a list operation and handles an AttributeError exception if the attribute does not exist.

```python
try:
    my_list = [1, 2, 3]
    my_list.push(4)
except AttributeError:
    print("Error: 'list' object has no attribute 'push'. Use 'append' instead.")
```

## Q3. Write a python program to handle multiple exception types in a single block.

```python
try:
    num = int(input("Enter a number: "))
    my_list = [10, 20, 30]
    
    result = 100 / num
    print(f"100 / {num} = {result}")
    
    print(f"Element at index {num}: {my_list[num]}")
except (ValueError, ZeroDivisionError, IndexError) as e:
    print(f"Error occurred: {e}")
```

## Q4. Write a Python program to demonstrate a user-defined exception that checks whether the entered marks are between 0 and 100. Display an appropriate message for invalid input and out-of-range values.

```python
class InvalidMarksException(Exception):
    pass

try:
    marks = int(input("Enter marks (0-100): "))
    
    if marks < 0 or marks > 100:
        raise InvalidMarksException("Marks must be between 0 and 100")
    
    print(f"Valid marks entered: {marks}")
except ValueError:
    print("Invalid input! Please enter a number.")
except InvalidMarksException as e:
    print(f"Error: {e}")
```
