## Q3. Write a python program to generate a random radius and computes the area (use math module).

```python
import math
import random

radius = round(random.uniform(1, 10), 2)
area = math.pi * radius ** 2

print("Radius:", radius)
print("Area:", round(area, 2))
```

## Q4. Write a Python program to get current time and convert to local time.

```python
import time

current = time.time()
local = time.localtime(current)

print("Current time:", current)
print("Local time:", time.strftime("%d-%m-%Y %H:%M:%S", local))
```

## Q7. Write a Python program to shuffle a deck of card. Display the cards obtained by a user in 5 attempts.

```python
import random

suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
ranks = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']

deck = [r + " of " + s for s in suits for r in ranks]
random.shuffle(deck)

for i in range(5):
    print(f"Card {i+1}: {deck[i]}")
```
