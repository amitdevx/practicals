# GroupBy Operations and Aggregation - Set C

```python
import pandas as pd
import numpy as np
```

## Q. Create a following DataFrame named as "data" and split the dataframe into groups based on school code.

```python
data = pd.DataFrame({
    'school': ['S1', 'S2', 'S3', 'S4', 'S5', 'S6'],
    'class': ['S001', 'S002', 'S003', 'S004', 'S005', 'S006'],
    'name': ['Asha', 'Fransis', 'Charlie', 'David', 'Shyam', 'Eashel'],
    'date_of_birth': ['15/05/2002', '17/05/2002', '16/02/1999', '25/09/1998', '11/05/2002', '15/09/1997'],
    'age': [12, 12, 13, 14, 12, 16],
    'height': [173, 192, 186, 167, 151, 159],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['Street1', 'Street2', 'Street3', 'Street4', 'Street5', 'Street6']
})

print("Original DataFrame")
print(data)
```

```python
grouped = data.groupby('school')

print("\nGrouped object created")
```

1. Check the type of GroupBy object.

```python
print("\n1. Type of GroupBy object:")
print(type(grouped))
```

2. Calculate mean, min, and max value of age for each school.

```python
print("\n2. Mean, Min and Max age for each school:")
print(grouped['age'].agg(['mean', 'min', 'max']))
```
