## Q4. Write a Python program to get current time and apply ctime(), gmtime(), strftime(), strptime().

```python
import time

t = time.time()

print("ctime:", time.ctime(t))
print("gmtime:", time.gmtime(t))
print("strftime:", time.strftime("%d-%m-%Y %H:%M:%S", time.localtime(t)))
print("strptime:", time.strptime("26-01-2024 11:30:00", "%d-%m-%Y %H:%M:%S"))
```
