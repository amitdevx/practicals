## Q1. Write a python program to check weather given url is valid or not? (for Ex: https://www.yahoo.com)

```python
def validate_url(url):
    pattern = r'^https?://[a-zA-Z0-9][-a-zA-Z0-9]*\.[a-zA-Z]{2,}(/.*)?$'
    if re.match(pattern, url):
        return True
    return False

urls = [
    'https://www.yahoo.com',
    'http://www.google.com',
    'https://github.com/amitdevx',
    'www.example.com',
    'invalid-url',
    'ftp://files.example.com'
]

for url in urls:
    print(f"URL '{url}' is valid: {validate_url(url)}")
```

## Q2. Write a python program to match word and single letter separated by a comma and single space as in first and last name.

```python
def match_name_pattern(string):
    pattern = r'^[A-Za-z]+, [A-Za-z]$'
    if re.match(pattern, string):
        return True
    return False

test_strings = [
    'Divekar, A',
    'Smith, J',
    'Kumar, R',
    'John,K',
    'A, Divekar',
    'Amit Divekar'
]

for s in test_strings:
    print(f"String '{s}' matches pattern: {match_name_pattern(s)}")
```

## Q3. Write a python program to accept a string, Count total number of vowels, consonents and special symbols from it.

```python
def count_characters(string):
    vowels = len(re.findall(r'[aeiouAEIOU]', string))
    consonants = len(re.findall(r'[bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ]', string))
    special_symbols = len(re.findall(r'[^a-zA-Z]', string))
    return vowels, consonants, special_symbols

text1 = "Hello, World! How are you?"
v, c, s = count_characters(text1)
print(f"String: {text1}")
print(f"Vowels: {v}")
print(f"Consonants: {c}")
print(f"Special Symbols: {s}")

print()

text2 = "Python Programming@2024"
v, c, s = count_characters(text2)
print(f"String: {text2}")
print(f"Vowels: {v}")
print(f"Consonants: {c}")
print(f"Special Symbols: {s}")
```

## Q4. Write a python program to accept a string, Find all words that do not contain vowels.

```python
def find_words_without_vowels(string):
    words = re.findall(r'\b[^aeiouAEIOU\s\d\W]*[bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ]+[^aeiouAEIOU\s\d\W]*\b', string)
    result = [word for word in words if not re.search(r'[aeiouAEIOU]', word)]
    return result

text1 = "My gym rhythm is fly by dry sky"
print(f"String: {text1}")
print(f"Words without vowels: {find_words_without_vowels(text1)}")

print()

text2 = "The cat sat on the dry mat by the gym"
print(f"String: {text2}")
print(f"Words without vowels: {find_words_without_vowels(text2)}")
```

## Q5. Write a python program to accept a string, Extract all words that start with a capital letter and remove all special characters from a string.

```python
def extract_capital_words(string):
    words = re.findall(r'\b[A-Z][a-zA-Z]*', string)
    return words

def remove_special_characters(string):
    result = re.sub(r'[^a-zA-Z0-9\s]', '', string)
    return result

text1 = "Hello, World! This is Python Programming@2024."
print(f"Original String: {text1}")
print(f"Words starting with capital letter: {extract_capital_words(text1)}")
print(f"String after removing special characters: {remove_special_characters(text1)}")

print()

text2 = "Amit Divekar from Mumbai, India! Studying Computer Science."
print(f"Original String: {text2}")
print(f"Words starting with capital letter: {extract_capital_words(text2)}")
print(f"String after removing special characters: {remove_special_characters(text2)}")
```
