# Regular Expression - Set A

```python
import re
```

## Q1. Write a python program to extract number from a given string.

```python
def extract_numbers(string):
    numbers = re.findall(r'\d+', string)
    return numbers

text1 = "My phone number is 9876543210 and my age is 21"
print(f"Original String: {text1}")
print(f"Extracted Numbers: {extract_numbers(text1)}")

print()

text2 = "There are 100 apples and 50 oranges"
print(f"Original String: {text2}")
print(f"Extracted Numbers: {extract_numbers(text2)}")
```

## Q2. Write a python program to remove all whitespaces from a given string.

```python
def remove_whitespaces(string):
    result = re.sub(r'\s+', '', string)
    return result

text1 = "Hello World   This is   Python Programming"
print(f"Original String: {text1}")
print(f"After removing whitespaces: {remove_whitespaces(text1)}")

print()

text2 = "   Amit   Divekar   "
print(f"Original String: {text2}")
print(f"After removing whitespaces: {remove_whitespaces(text2)}")
```

## Q3. Write a python program to check if a particular substring is present at the beginning of the given string.

```python
def check_beginning(string, substring):
    pattern = r'^' + re.escape(substring)
    if re.search(pattern, string):
        return True
    return False

text1 = "Hello World"
print(f"String: {text1}")
print(f"Substring 'Hello' is present at the beginning: {check_beginning(text1, 'Hello')}")
print(f"Substring 'World' is present at the beginning: {check_beginning(text1, 'World')}")

print()

text2 = "Python Programming"
print(f"String: {text2}")
print(f"Substring 'Python' is present at the beginning: {check_beginning(text2, 'Python')}")
print(f"Substring 'Java' is present at the beginning: {check_beginning(text2, 'Java')}")
```

## Q4. Write a python program to validate strings like bit, but, bat, hit, hut, hat.

```python
def validate_string(string):
    pattern = r'^[bh][aiu]t$'
    if re.match(pattern, string):
        return True
    return False

test_strings = ['bit', 'but', 'bat', 'hit', 'hut', 'hat', 'cat', 'bot', 'bits']

for s in test_strings:
    print(f"String '{s}' is valid: {validate_string(s)}")
```

## Q5. Write a Python program to replace all occurrences of a space, comma, or dot with a colon.

```python
def replace_with_colon(string):
    result = re.sub(r'[ ,.]', ':', string)
    return result

text1 = "Python Exercises, Data Structure exercises."
print(f"Input: {text1}")
print(f"Output: {replace_with_colon(text1)}")

print()

text2 = "Hello, World. How are you?"
print(f"Input: {text2}")
print(f"Output: {replace_with_colon(text2)}")
```
