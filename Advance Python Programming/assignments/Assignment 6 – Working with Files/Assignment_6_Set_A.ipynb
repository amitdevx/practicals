# Working with Files - Set A

## Q1. Write a Python program to read an entire text file line by line.

```python
with open("sample.txt", "w") as f:
    f.write("Hello World\nPython Programming\nFile Handling")

with open("sample.txt", "r") as f:
    for i, line in enumerate(f, 1):
        print(f"Line {i}: {line.strip()}")
```

## Q2. Write a Python program to store the names of 5 friends in a text file.

```python
friends = ["Rahul", "Amit", "Priya", "Sneha", "Rohan"]

with open("friends.txt", "w") as f:
    for friend in friends:
        f.write(friend + "\n")

print("Friends saved to file!")

with open("friends.txt", "r") as f:
    print(f.read())
```

## Q3. Write a Python program to append text to an existing file and display the updated content.

```python
with open("append_demo.txt", "w") as f:
    f.write("Hello World\n")

print("Original content:")
with open("append_demo.txt", "r") as f:
    print(f.read())

with open("append_demo.txt", "a") as f:
    f.write("This is appended text.\n")

print("After appending:")
with open("append_demo.txt", "r") as f:
    print(f.read())
```

## Q4. Write a Python program to get the file size, creation time, last modification time, and last access time of a file.

```python
import os
import time

filename = "sample.txt"
stat = os.stat(filename)

print(f"File: {filename}")
print(f"Size: {stat.st_size} bytes")
print(f"Created: {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(stat.st_ctime))}")
print(f"Modified: {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(stat.st_mtime))}")
print(f"Accessed: {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(stat.st_atime))}")
```

## Q5. Write a Python program to demonstrate the use of tell() and seek() methods.

```python
with open("seek_demo.txt", "w") as f:
    f.write("Hello World")

with open("seek_demo.txt", "r") as f:
    print(f"Initial position: {f.tell()}")
    
    print(f"After reading 5 chars: {f.read(5)}")
    print(f"Current position: {f.tell()}")
    
    f.seek(0)
    print(f"After seek(0): {f.read()}")
    
    f.seek(6)
    print(f"After seek(6): {f.read()}")
```

## Q6. Write a Python program to list only files in a directory using os.listdir().

```python
import os

print("Files in current directory:")
for item in os.listdir("."):
    if os.path.isfile(item):
        print(f"- {item}")
```
