## Q2. Write a Python program to find all .txt files in a directory and its subdirectories using os.walk().

```python
import os

os.makedirs("walk_demo/subdir", exist_ok=True)
open("walk_demo/file1.txt", "w").close()
open("walk_demo/subdir/file2.txt", "w").close()
open("walk_demo/script.py", "w").close()

print("All .txt files:")
for root, dirs, files in os.walk("walk_demo"):
    for file in files:
        if file.endswith(".txt"):
            print(f"- {os.path.join(root, file)}")
```

## Q3. Write a Python program to generate a file report containing name, size, and modification date.

```python
import os
import time

os.makedirs("report_dir", exist_ok=True)
with open("report_dir/report1.txt", "w") as f:
    f.write("First report content")
with open("report_dir/report2.txt", "w") as f:
    f.write("Second report content here")

print("File Report:")
print("-" * 40)
for filename in os.listdir("report_dir"):
    filepath = os.path.join("report_dir", filename)
    if os.path.isfile(filepath):
        stat = os.stat(filepath)
        print(f"Name: {filename}")
        print(f"Size: {stat.st_size} bytes")
        print(f"Modified: {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(stat.st_mtime))}")
        print("-" * 40)
```

## Q4. Write a Python program to scan a directory and separate files by type.

```python
import os

os.makedirs("scan_dir", exist_ok=True)
open("scan_dir/doc1.txt", "w").close()
open("scan_dir/doc2.txt", "w").close()
open("scan_dir/script.py", "w").close()
open("scan_dir/image.jpg", "w").close()

files_by_type = {}
for filename in os.listdir("scan_dir"):
    if os.path.isfile(os.path.join("scan_dir", filename)):
        ext = os.path.splitext(filename)[1]
        if ext not in files_by_type:
            files_by_type[ext] = []
        files_by_type[ext].append(filename)

print("Files by type:")
for ext, files in files_by_type.items():
    print(f"{ext} files: {files}")
```

## Q5. Write a Python program to rename a file by adding the current date to its name.

```python
import os
import time

os.makedirs("rename_dir", exist_ok=True)
original = "rename_dir/original.txt"
with open(original, "w") as f:
    f.write("Content")

date_str = time.strftime("%Y-%m-%d")
name, ext = os.path.splitext(original)
new_name = f"{name}_{date_str}{ext}"

os.rename(original, new_name)
print(f"Original: {os.path.basename(original)}")
print(f"Renamed: {os.path.basename(new_name)}")
```

## Q6. Write a Python program to create a backup of a directory using shutil.copytree().

```python
import shutil
import os

os.makedirs("backup_src", exist_ok=True)
with open("backup_src/file1.txt", "w") as f:
    f.write("File 1 content")
with open("backup_src/file2.txt", "w") as f:
    f.write("File 2 content")

if os.path.exists("backup_dest"):
    shutil.rmtree("backup_dest")

shutil.copytree("backup_src", "backup_dest")
print("Backup created successfully!")

print("Backup contents:")
for f in os.listdir("backup_dest"):
    print(f"- {f}")
```
