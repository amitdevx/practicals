## Q1. Write a Python program to count the number of characters, words, and lines in a text file.

```python
with open("count_demo.txt", "w") as f:
    f.write("Hello World\nPython is awesome\nFile handling in Python")

with open("count_demo.txt", "r") as f:
    content = f.read()
    chars = len(content)
    words = len(content.split())
    lines = content.count("\n") + 1

print(f"Characters: {chars}")
print(f"Words: {words}")
print(f"Lines: {lines}")
```

## Q2. Write a Python program to copy the contents of one file to another.

```python
with open("source.txt", "w") as f:
    f.write("This is source file content.\nLine 2 of source.")

with open("source.txt", "r") as src:
    with open("destination.txt", "w") as dest:
        dest.write(src.read())

print("File copied successfully!")
print("Destination content:")
with open("destination.txt", "r") as f:
    print(f.read())
```

## Q3. Write a Python program to count the number of .py files in a directory.

```python
import os

os.makedirs("py_test", exist_ok=True)
open("py_test/test1.py", "w").close()
open("py_test/test2.py", "w").close()
open("py_test/readme.txt", "w").close()

py_files = [f for f in os.listdir("py_test") if f.endswith(".py")]
print(f"Number of .py files: {len(py_files)}")
for f in py_files:
    print(f"- {f}")
```

## Q4. Write a Python program to identify whether a given path is a file or folder.

```python
import os

paths = ["count_demo.txt", "py_test", "nonexistent"]

for path in paths:
    if os.path.isfile(path):
        print(f"'{path}' is a File")
    elif os.path.isdir(path):
        print(f"'{path}' is a Directory")
    else:
        print(f"'{path}' does not exist")
```

## Q5. Write a Python program to read a file from a specific position using seek().

```python
with open("seek_test.txt", "w") as f:
    f.write("Python Programming Language")

with open("seek_test.txt", "r") as f:
    print(f"Full content: {f.read()}")
    
    f.seek(7)
    print(f"From position 7: {f.read()}")
    
    f.seek(19)
    print(f"From position 19: {f.read()}")
```
